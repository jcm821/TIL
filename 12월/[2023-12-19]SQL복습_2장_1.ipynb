# SQL로 맛보는 데이터 전처리 분석

# 2장. SQL문법

```python
# 컬럼 조회
SELECT 호출하려는 컬럼
FROM DB명.테이블 명;
```

```python
# 요청사항: classicmodels.customers의 customerNumber를 조회하세요
USE classicmodels
SELECT customerNumber FROM customers;
```

```python
# 컬럼의 개수 조회
SELECT COUNT(상품 번호)
FROM DB명.PRODUCT;
```

```python
# 요청사항: 상품 개수도 조회가능할까요?
SELECT COUNT(productCode) FROM products;
```

```python
# 집계 함수
SELECT 집계함수
FROM DB명.PRODUCT;
```

```python
# 요청사항: classicmodels.payments의 amount의 총합과 checknumber 개수를 구하시오
SELECT SUM(amount), COUNT(checknumber)
FROM payments;
```

```python
# 모든 결과 조회 -> * 연산자 이용
SELECT *
FROM DB명.테이블 명;
```

```python
# 여러 개의 컬럼을 출력
SELECT 컬럼 명1, 컬럼 명2
FROM DB명.테이블 명;
```

```python
# 요청사항: classicmodels.products의 productName, productLine을 조회하세요
SELECT productName, productLine
FROM products;
```

```python
# 컬럼명 변결: AS 이용
SELECT 컬럼 명1 as 변경 컬럼 명
FROM DB명.테이블 명;
```

```python
# 요청사항: products의 productCode의 개수를 구하고, 컬럼 명을 n_products로 변경하세요
# as를 생략하고 변수명을 변경해도 가능하다
SELECT COUNT(productCode) as n_products
FROM products;
```

```python
# 중복을 제외하고 데이터 조회: DISTINCT
SELECT DISTINCT 제조국가
FROM DB명.상품 테이블;
```

```python
# 요청사항: orderdetails의 ordernumber의 중복을 제거하고 조회하세요
SELECT DISTINCT ordernumber
FROM orderdetails;
```

```python
# 조건절: WHERE
# WHERE를 추가하여 조건을 생성할 수 있다
SELECT 상품번호
FROM DB명.PRODUCT
WHERE 판매국가 = '미국';
```

```python
# BETWEEN: 특정 컬럼의 값이 시작점 ~ 끝점인 데이터만 출력
SELECT *
FROM DB명.테이블 명
WHERE 컬럼 BETWEEN 시작점 AND 끝점;
```

```python
# 출시일이 2010~2014년인 조건 추가
SELECT 상품번호
FROM DB명.PRODUCT
WHERE 출시연도 BETWEEN 2010 AND 2014;
```

```python
# 요청사항: orderdetails의 priceeach가 30에서 50사이인 데이터를 조회하시오
SELECT *
FROM orderdetails
WHERE priceeach BETWEEN 30 AND 50;
```

```python
# 다음과 같이 대소관계 연산자를 사용할 수 있다
SELECT 상품번호
FROM DB명.PRODUCT
WHERE 출시연도 < 2010;

SELECT 상품번호
FROM DB명.PRODUCT
WHERE 출시연도 > 2014;
```

```python
# 요청사항: orderdetails의 priceeach가 30이상인 데이터를 조회하세요
SELECT *
FROM orderdetails
WHERE priceeach >= 30;
```

```python
# IN연산자를 사용하여 여러 조건을 포함시킬 수 있다.
SELECT 컬럼 명
FROM 테이블 명
WHERE 컬럼 명 IN (값1, 값2);
```

```python
# 요청사항: customers의 country가 USA 또는 Canada인 customernumber를 조회하세요
SELECT customernumber
FROM customers
WHERE country IN ('USA', 'Canada');
```

```python
# NOT IN 연산자를 이용해서 특정 값들을 제외한 결과를 출력할 수 있다.
SELECT 상품번호
FROM DB명.테이블 명
WHERE 판매국가 NOT IN ('미국', '영국');
```

```python
# 요청사항: customers의 country가 USA, Canada가 아닌 customernumber를 조회하세요
SELECT customernumber
FROM customers
WHERE country NOT IN ('USA', 'Canada');
```

```python
# NULL데이터 출력: IS NULL
SELECT 컬럼 명 또는 계산식
FROM 테이블 명
WHERE 컬럼 IS NULL;
```

```python
# 요청사항: employees의 reportsTo 값이 NULL인 employeenumber를 조회하세요
SELECT employeenumber
FROM employees
WHERE reportsTo IS NULL;
```

```python
# LIKE연산자: 특정 필드에 어떤 텍스트가 포함되는 경우를 추출
SELECT *
FROM DB명.CUSTOMERS
WHERE 주소 LIKE '%부산%';
```

```python
# 요청사항: customers의 addressline1에 ST가 포함된 addressline1을 출력하세요
SELECT addressline1
FROM customers
WHERE addressline1 LIKE '%ST%';
```
