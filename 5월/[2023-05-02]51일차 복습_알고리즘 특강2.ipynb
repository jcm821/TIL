* 위와 같이 앞에 데이터가 비어있음에도 불구하고 큐가 꽉차있다는 구조적인 오류를 발견할 수 있다. 따라서 예외처리를 위해 isQueueFull()함수의 case조건을 확장시켜 줄 필요가 있다.

```python
def isQueueFull():
    global SIZE, queue, front, rear
    # 일반화를 위해 케이스를 증가시켜준다
    
    # Case1. 'rear != 큐 크기 - 1'인 경우
    if (rear != SIZE - 1):
        return False
    
    # Case2. 'rear = 큐 크기 - 1 이고 front = -1'일 경우(완전히 꽉 찬 경우)
    elif (rear == SIZE - 1 and front == -1):
        return True
    
    # Case3. 'rear = 큐 크기 - 1 이고 front != -1'일 경우(앞에 자리가 난 경우)
    elif (rear == SIZE - 1 and front != -1):
        for i in range(front + 1, SIZE, 1):
            # front + 1 위치부터 rear위치까지 왼쪽으로 한 칸씩 이동
            queue[i - 1] = queue[i]
            queue[i] = None
        # front와 rear 값에서 1을 뺀다
        front -= 1
        rear -= 1
        return False
```

### 원형 큐

* 큐의 처음과 끝이 연결되어있는 구조이다.
* 데이터가 방대할 경우, 순차큐에서는 오버플로우가 발생할 가능성이 높다.
* 이를 방지하기 위하여 순차 큐를 구부려서 이은 형태의 원형 큐로 변형하게 된다면, 오버헤드가 발생하지 않게 된다.
* 원형 큐가 빈 경우: front와 rear가 동일할 때 큐가 빈것으로 처리
* 원형 큐가 꽉 찬 경우: rear + 1과 front가 같은 경우에 원형 큐가 꽉 찬 것으로 처리
    - rear와 front가 같아버리면 큐가 빈 지, 꽉 찬 것인지 확인하기 힘드므로 해석에 오류 발생

```python
## 원형 큐

# 함수

# 큐가 꽉 찾는지 확인하는 함수(오버플로우 방지)
def isQueueFull():
    global SIZE, queue, front, rear
    if ((rear + 1) % SIZE == front):
        return True
    else:
        return False
def enQueue(data):
    global SIZE, queue, front, rear
    if (isQueueFull()):
        print('큐가 꽉찼습니다')
        return
    rear = (rear + 1) % SIZE
    queue[rear] = data

# 큐가 비어있는지 확인하는 함수(언더플로우 방지)
def isQueueEmpty():
    global SIZE, queue, front, rear
    if (front == rear):
        return True
    else:
        return False


def deQueue():
    global SIZE, queue, front, rear
    if (isQueueEmpty()):
        print('손님이 없습니다')
        return
    front = (front + 1) % SIZE
    data = queue[front]
    queue[front] = None
    return data

# 데이터를 확인하는 함수
def peek():
    global SIZE, queue, front, rear
    if (isQueueEmpty()):
        print('손님이 없습니다')
        return
    return queue[(front + 1) % SIZE]

# 변수
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = 0
```

## 2. 이진트리

### 트리구조
* 조직표, 컴퓨터의 상위 폴더 안에 하위폴더들이 계속 이어져 있는 구조와 같은 구성

### 이진 트리
* 트리(Tree) 자료구조: 나무를 거꾸로 뒤집어 놓은 형태
* 차수: 자식노드의 개수

### 이진 트리의 종류
* 포화 이진 트리, 완전 이진 트리, 편향 이진 트리

### 이진 트리의 순회
* 순회 종류
    - 이진 트리의 노드 전체를 한 번씩 방문하는 것을 순회(traversal)라고 한다.
    - 노드 데이터를 처리하는 순서에 따라 전위 순회, 중위 순회, 후위 순회

### 이진 탐색 트리의 특징
이진 트리 중 활용도가 높은 트리, 데이터 크기를 기준으로 일정 형태로 구성한다.
1. 왼쪽 서브 트리는 루트 노드보다 모두 작은 값을 가진다.
2. 오른쪽 서브 트리는 루트 노드보다 모두 큰 값을 가진다.
3. 각 서브 트리도 1, 2의 특징을 갖는다.
4. 모든 노드 값은 중복되지 않는다. 즉, 중복된 값은 이진 탐색트리에 저장할 수 없다.

```python
# 기초적인 이진트리
# 함수
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

# 전역

# 메인
# 이진트리 생성
node1 = TreeNode()
node1.data = '화사'

node2 = TreeNode()
node2.data = '솔라'
node1.left = node2

node3 = TreeNode()
node3.data = '문별'
node1.right = node3

node4 = TreeNode()
node4.data = '휘인'
node2.left = node4

node5 = TreeNode()
node5.data = '쯔위'
node2.right = node5

node6 = TreeNode()
node6.data = '선미'
node3.left = node6

root = node1
# 출력
print(root.data)
print(root.left.data, root.right.data)
print(root.left.left.data, root.left.right.data, root.right.left.data)
```

#### 가중치 그래프
* 간선마다 가중치가 다르게 부여된 그래프
* 무방향 그래프와 방향 그래프에 각각 가중치를 부여한 경우 다음의 표와 같다.

* 무방향 가중치 그래프



* 방향 가중치 그래프


### 깊이 우선 탐색의 작동
* 그래프의 모든 정점을 한 번 씩 방문하는 것을 순회(Graph Traversal)라고 한다.
* 순회방식으로 깊이 우선 탐색, 너비 우선 탐색이 대표적이다.

### 그래프의 인접 행렬 표현
* 그래프를 코드로 구현할 때는 인접 행렬 사용
* 인접 행렬은 정방형으로 구성된 행렬로 정점이 4개인 그래프는 4x4로 표현한다.

```python
# 그래프
# 함수
class Graph():
    def __init__(self, size):
        self.graph = [[0 for _ in range(size)] for _ in range(size)]

# 변수
A, B, C, D = 0, 1, 2, 3
# 메인
G1 = Graph(4)
G1.graph[A][B] = 1
G1.graph[A][C] = 1
G1.graph[A][D] = 1

G1.graph[B][A] = 1
G1.graph[B][C] = 1
G1.graph[B][D] = 1

G1.graph[C][A] = 1
G1.graph[C][B] = 1
G1.graph[C][D] = 1

G1.graph[D][A] = 1
G1.graph[D][C] = 1

for row in range(4):
    for col in range(4):
        print(G1.graph[row][col], end = ' ')
    print()
```

## 4. 정렬 알고리즘

* 순서대로 데이터가 나열되어 있는 것
* 중요 알고리즘 중 하나인 정렬(Sort)은 자료들을 순서대로 나열한 것이다.

### 정렬 알고리즘의 종류
* 오름차순 정렬이든 내림차순 정렬이든 결과의 형태만 다를 뿐이지 같은 방식으로 처리된다.
* 정렬하는 방법에 대한 정렬 알고리즘은 수십가지나 될 정도로 많다.

    - 선택정렬(Selection Sort)
    - 삽입정렬(Insertion Sort)
    - 버블정렬(Bubble Sort)
    - 퀵 정렬(Quick Sort)

### 선택정렬
* 여러 데이터 중에서 가장 작은 값을 뽑는 작동을 반복하여 값을 정렬하는 알고리즘

#### 최솟값을 찾는 방법
1. 배열의 첫 번째 값을 가장 작은 값으로 지정한다.
2. 가장 작은 값으로 지정한 값을 다음 차례의 값과 비교하여 가장 작은 값을 변경하거나 그대로 둔다.
3. 마지막 값까지 비교를 마친 후 현재 가장 작은 값으로 지정된 값을 가장 작은 값으로 결정한다.

```python
## 기초적인 정렬
## 선택 정렬
## 최소값 찾기

# 함수
import random
def findMinIndex(ary):
    minIdx = 0
    for i in range(1, len(ary)):
        if (ary[minIdx] > ary[i]):
            minIdx = i
    return minIdx




# 전역
# testAry = [55, 88, 33, 77]
testAry = [random.randint(30, 200) for _ in range(20)]
# 메인
print(testAry)
minPos = findMinIndex(testAry)
print('최솟값-->', testAry[minPos])
```

```python
# 선택정렬의 일반화

# 함수
import random
def findMinIndex(ary):
    minIdx = 0
    for i in range(1, len(ary)):
        if (ary[minIdx] > ary[i]):
            minIdx = i
    return minIdx
# 변수
before = [random.randint(30, 200) for _ in range(10)]
after = []

# 메인
print('정렬 전 -->', before)
for _ in range(len(before)):
    minPos = findMinIndex(before)
    after.append(before[minPos])
    del (before[minPos])
    
print('정렬 후 -->', after)
```

* 개선된 선택 정렬 알고리즘은 아래와 같다.

```python
# 개선된 선택 정렬
# 변수
dataAry = [random.randint(30, 200) for _ in range(10)]
print('정렬 전 -->', dataAry)
```

```python
# 함수
import random
def selectSort(ary):
    n = len(ary)
    # 사이클 순회 설정
    for i in range(0, n - 1):
        minIdx = i
        # 반복을 위한 순회
        for k in range(i + 1, n):
            if (ary[minIdx] > ary[i]):
                minIdx = k
        ary[i], ary[minIdx] = ary[minIdx], ary[i]
    return ary

dataAry = selectSort(dataAry)
print('정렬 후 -->', dataAry)
```

## 5. 검색 알고리즘

* '검색'은 정렬된 상태에서 빠르게 원하는 것을 찾을 수 있다.
* 어떤 집합에서 원하는 것을 찾는 것으로, 탐색이라고도 한다.
* 검색에는 순차 검색, 이진 검색, 트리 검색 등이 있다.
* 검색에 실패하면 -1을 반환하는 것이 일반적임

### 검색 알고리즘의 종류
#### 순차 검색
* 검색할 집합이 정렬되어 있지 않은 상태일 때
* 처음부터 차례대로 찾아보는 것으로, 쉽지만 비효율적
* 집합의 데이터가 정렬되어 있지 않다면 이 검색 외에 특별한 방법은 없다.

#### 이진 검색
* 데이터가 정렬되어 있다면 이진 검색도 사용 가능하다.
* 순차 검색에 비해 월등히 효율적이라 데이터가 몇 천만 개 이상이어도 빠르게 찾아낼 수 있다.

#### 트리 검색
* 데이터 검색에는 상당히 효율적이지만 트리의 삽입, 삭제 등에는 부담

```python
# 순차 검색

# 함수
import random
def seqSearch(ary, fData):
    pos = -1
    for i in range(len(ary)):
        if (ary[i] == findData):
            pos = i
            break
            
    return pos

# 변수
dataAry = [random.randint(30, 200) for _ in range(8)]
findData = random.choice(dataAry)

# 메인
print('배열 -->', dataAry)
position = seqSearch(dataAry, findData)
if (position == -1):
    print(findData, '는 없어요.')
else:
    print(findData, '는', position, '번째 있어요.')
```

```python
# 이진 검색

# 함수
import random
def binSearch(ary, fData):
    global count
    pos = -1
    start = 0
    end = len(ary) - 1
    while (start <= end):
        count += 1
        mid = (start + end) // 2
        if (ary[mid] == fData):
            pos = mid
            break
        elif (ary[mid] < fData):
            start = mid + 1
        else:
            end = mid - 1
    return pos

# 변수
# 성능확인을 위한 count변수 선언
count = 0
dataAry = [random.randint(0, 1000000) for _ in range(100000)]
findData = random.choice(dataAry)
dataAry.sort()
# 메인
print('배열 -->', dataAry[:20])
position = binSearch(dataAry, findData)
if (position == -1):
    print(findData, '는 없어요.')
else:
    print(findData, '는', position, '번째 있어요.', count)
```

## 6. 재귀 알고리즘

* 동일한 작동을 무한적으로 반복하는 알고리즘
* 재귀 호출(Recursion)은 자신을 다시 호출하는 것이다.

```python
# 잘못된 재귀함수
def openBox():
    print('상자를 엽니다')
    openBox()
openBox()
```

* 위의 결과를 통해 알 수 있듯, 잘못된 재귀함수는 무한루프를 돌게 되어 오류가 발생하게 된다.

```python
# 재귀 알고리즘의 기초
def openBox():
    global count
    print('상자를 엽니다')
    count -= 1
    if count == 0:
        print('선물 넣기')
        return
    openBox()
    print('상자 닫습니다')
    
count = 10
openBox()
```

* for문을 사용하지 않고 재귀를 통해 다음과 같이 구현이 가능하다.

```python
# 합계를 구하는 재귀
def addNumber(num):
    if num == 1:
        return 1
    return num + addNumber(num - 1)

print(addNumber(10))
```

```python
# 팩토리얼 재귀
def factorial(num):
    if num == 1:
        return 1
    return num * factorial(num - 1)
print(factorial(5))
```

```python
# 우주선 발사 카운트
def countDown(n):
    if n == 0:
        print('발사')
    else:
        print(n)
        countDown(n - 1)
countDown(5)
```

```python
# 별모양 출력하기
def printStar(n):
    if n > 0:
        printStar(n - 1)
        print('*' * n)
        
printStar(5)
```

```python
# 구구단 출력하기
def guguDan(dan, num):
    print('{} X {} = {}'.format(dan, num, dan * num))
    if num < 9:
        guguDan(dan, num + 1)
        
for dan in range(2, 10):
    print('## {}단 ##'.format(dan))
    guguDan(dan, 1)
```
