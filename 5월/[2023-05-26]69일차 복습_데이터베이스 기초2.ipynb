# 1. 데이터 품질관리

## 데이터(Data)

* 특정 목적의 활동이나 이벤트로 인해 발생된 사실이나 바탕이 되는 자료
* 광의의 개념으로 정보 또는 지식과 혼용하여 사용

## 데이터품질(Data Quality)

* 데이터의 최신성, 정확성, 상호연계성 등을 확보하여 사용자에게 유용한 가치를 줄 수 있는 수준

## 데이터 품질관리(Data Quality Management)

* 데이터품질을 확보하기 위한 일련의 활동
* 일련의 활동: 품질목표 설정, 품질진단, 품질개선, 지원도구 제공 등

### 정보생명주기
* 계획 -> 획득 -> 저장/공유 -> 운영 -> 활용 -> 폐기
* 일반적으로 데이터 품질관리는 운영 및 활용 단계에서 이뤄지는 업무로 인식
* 공공데이터 품질관리는 운영 및 활용 단계에서 이뤄지는 업무로 인식
* 품질관리 활동은 정보생명주기를 고려하여 계획 -> 구축 -> 운영 -> 활용 4단계로 구성

# 5. 데이터 표준화

* 기관 내 시스템별로 산재되어 있는 데이터 정보 요소에 대하여 명칭, 정의, 형식, 규칙 등에 관한 원칙을 수립하고, 이를 기관 전체의 데이터에 적용하는 것
* 개별 공공기관들은 범정부 공통표준용어를 바탕으로 데이터 표준화 시행
* 적용대상: 공공데이터베이스, 즉 공공기관이 구축/운영/관리(업무위임/위탁관리를 포함)하는 데이터베이스
* 적용시점: <br>신규로 공공데이터베이스를 구축할 때 -> 바로적용<br>기존 운영 중인 공공데이터베이스 -> 신규 전면 구축 또는 재구축 업무 추진 시 적용

### 공통표준단어 표준화
* 여러 기관에서 상이하게 사용하는 단어들 가운데 범정부 차원에서 대표가 되는 표준단어를 선정하고 관리항목을 정의하는 것

### 단어
* 분리하여 자립적인 의미를 전달할 수 있는 말의 최소 단위

### 업무단어
* 단어 중에서 업무의 의미를 파악할 수 있는 단어, 용어를 구성할 때 형식단어 앞에 위치(수식어)

### 형식단어
* 단어 중에서 데이터값의 저장/표현형식, 길이 등 특정한 형식 식별이 가능한 단어, 용어 구성 시 업무단어 뒤에 위치(분류어)

### 공통표준도메인 표준화
* 여러 공공기관에서 사용하는 공통표준용어의 데이터 속성을 분석하여 도메인 분석체계를 수립하고 각 구성요소를 정의한 것으로 모든 기관에 공통적으로 적용할 수 있는 상위 수준에서의 정의

### 도메인
* 데이터 값의 공통으로 갖는 데이터 형식과 허용되는 값의 영역을 정의한 것

* 기관별 상이한 업무특성으로 인해 데이터 속성에 대한 일관된 표준을 제시하기에 한계가 있으므로 범정부적으로 공동 활용할 공통표준용어 형식의 일관성을 확보하기 위해 추진
* 용어 끝의 형식단어(분류어)를 사용하여 정의

### 공통표준용어 표준화
* 데이터 사용자간의 명확한 의사소통을 지원하기 위해 여러 기관에서 공통적으로 사용하는 용어들 가운데 공통표준용어를 정하고 생성/관리 원칙을 수립하는 것

### 용어
* 단어 및 단어의 조합으로 이루어진 말, 업무단어(수식어) + 형식단어(분류어)의 형태
    - 기관표준용어: 각 기관이 구축/운영하는 데이터베이스에 일관된 용어를 적용하여 기관 내부에서 공동 활용하기 위해 정한 용어의 표준
    - 분야별표준용어: 동일 업무분야 내에서 공공데이터를 공동 활용하기 위하여 정한 용어의 표준

* 기관별로 상이하게 사용되는 동일의미 유사컬럼들에 대해 공통 표준 용어를 적용함으로써 컬럼을 용이하게 식별하고 다기관 데이터 통계산출 및 융합/분석에 일관된 기준 제공
* 개별기관에서 신규 요건 발생으로 용어 작명 시 신속하게 공통표준용어를 참조/활용할 수 있어 시간 단축, 업무생산성 향상 가능
