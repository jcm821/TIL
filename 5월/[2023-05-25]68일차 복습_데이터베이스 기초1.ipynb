# 1. 데이터, 정보, 지식

### 데이터
* 가공되지 않은 상태의 단순 사실
* 현실세계로부터 단순한 관찰이나 측정을 통해서 수집된 사실(fact)이나 값(value)
* 숫자, 스트링, 텍스트, 이미지, 그래픽스
* 대부분 운영 데이터베이스에 수집해 놓은 상태 값

### 정보
* 데이터에 의미있는 가공 절차를 적용한 산출물
* 어떤 상황에 대한 의사결정을 할 수 있게 하는 지식, 데이터의 유효한 해석이나 데이터 상호간 관계
* 정화되고 처리된 데이터

### 지식
* 목적 지향적으로 추상화/일반화된 가공 정보
* 의사결정과 비즈니스 이해를 위한 기반

# 2. 정보시스템

## 정의

* 한 조직체의 활동에 필요한 데이터를 수집, 조직, 저장해 두었다가 필요 시에 처리해서 의사결정에 유효한 정보를 생성 분배하는 수단
* 한 조직체의 내부적 운영과 외부적 상황에 관련된 과거, 현재, 그리고 예측되는 미래상황에 대한 정보를 체계적으로 제공하는 방법

### 일괄처리 시스템
* 응용의 특성상 데이터를 수집해서 분류하고 정렬시킨 다음, 일괄처리하는 데이터 처리 방식의 시스템
* 입시처리, 성적처리, 급여처리 시스템

### 온라인 처리 시스템
* 데이터는 생성되는 출처로부터 곧바로 컴퓨터에 전송되고, 전송된 데이터는 컴퓨터가 즉시 처리해서 그 결과를 원하는 장소로 보내주는 시스템
* 오늘날의 거의 모든데이터 처리 업무
* 항공 및 철도 좌석 예약 업무 처리 시스템 등

### 분산 처리 시스템
* 지리적(물리적)으로 분산되어 있는 처리기와 데이터베이스를 네트워크를 연결시켜 마치 논리적으로 하나의 시스템을 사용하는 것처럼 데이터를 처리헤주는 시스템

# 3. 데이터베이스

## 정의

* 한 조직이 여러 응용시스템이 공용하기 위해 최소의 중복으로 통합(Integrated), 저장(Stored)된, 운영(Operational) 데이터의 집합

|정의요소|내용|핵심|
|----|----|----|
|통합된 데이터|- 똑같은 데이터가 원칙적으로 중복되어 있지 않다는 것 의미 <br> - 최소의 중복(Minimal Redundancy) or 통제된 중복(Controlled Redundancy)|최소중복|
|저장된 데이터|- 컴퓨터가 접근할 수 있는 저장 매체에 저장된 데이터|컴퓨터접근가능|
|운영 데이터|- 어떤 조직의 고유 기능 수행을 위한 데이터|기능수행|
|공용 데이터|- 조직에 있는 여러 응용 시스템들이 공동으로 생성하고 유지하며 이용하는 공동의 데이터|공동 이용|

## 특징

|주요 특징|설명|기술요소|
|----|----|----|
|실시간 접근|- 사용자의 Query에 대해 실시간 온라인 처리로 응답|- 응답시간, 튜닝|
|계속적 변화|- 데이터 삽입, 삭제, 갱신(Insert/delete/update)작업이 수시로 발생|- SQL|
|내용에 의한 참조|- 데이터 레코드의 주소/위치가 아닌, 데이터 내용(값)에 따라 참조|- 물리적독립성 <br> - Index
|동시 공용|- 여러 사용자가 동시에 자기가 원하는 데이터에 접근 가능|- 동시성제어(Serialization)|


# 4. 파일시스템의 문제점 및 특성

* 데이터 일관성 보장 문제
* 데이터 무결성 보장 문제
* 공유성, 사용편의성 저조
* 프로그램 의존적
* 의미적으로 같은 값은 동일하게 유지되어야 함
* 낮은 경제성, 보안관리 저조

## 데이터베이스 시스템 장단점

|장점|단점|
|---|---|
|- 자료독립성 <br>- 중복자료감소 <br>- 자료통합처리 <br> - 자체 자료관리기능(병행제어/회복/무결성/보안) <br> - 생산성 향상(표준화도구/개발기간단축/유지보수 비용절감)|- 시스템 부담증가로 인한 저효율 <br> - 시스템 복잡합 <br> - 복잡한 연산 위한 고급 인력 필요 <br> - 어려운 백업 및 회복 <br> - 높은 비용 및 고급인력필요|

## 개념

* 응용프로그램과 Database 사이의 중재자로서 모든 응용프로그램들이 Data를 공유할 수 있게 관리해주는 소프트웨어 시스템

# 8. 관계형 DBMS(RDBMS, Relational DBMS)

## 정의

* 1970년대 관계형 데이터모델에 기반한 데이터베이스 관리 시스템

|특징|설명|
|----|---|
|모델 자체가 간단함|이차원 구조(열,행)에 문자/숫자/날짜 정보 저장하는 형태|
|수학적 이론의 바탕|수학의 집합론 바탕으로 하고 있어 개발할 시스템의 성능을 수학적으로 미리 예측 검증할 수 있으며 여러 연산을 수학적으로 최적화 할 수 있음|
|질의어 존재|간단한 질의어만 익히면 누구나 원하는 정보를 쉽게 검색할 수 있음|
|기술의 지속적 지원|시대적상황에 부응하여 C/S구조, 대규모 별렬처리의 지원|


# 9. 객체형 DBMS(OODBMS, Object-Oriented DBMS)

## 정의

* 객체모델에 기반하여 정보의 저장 및 검색을 지원해주는 데이터베이스

|등장배경|설명|
|---|---|
|Impedance Mismatch|프로그래밍 언어 지원 TYPE과 DB 지원 TYPE의 차이로 인하여 프로그램에서 인위적으로 변환 과정을 거치는 현상 및 문제점|
|객체지향 설계 요구의 증가|재사용을 극대화 하기 위해 객체 설계에 대한 요구 증가|
|관계형 모델의 한계|이미지, Text, 오디오, 비디오등 복잡한 모델링의 한계|



## 특징

|특징|설명|
|---|---|
|사용자정의타입 지원|사용자 정의 타입지원과 이들간의 상속성 명세 가능|
|비정형모델링|비정형 복합정보의 모델링 기능|
|네비게이션 기반 정보접근|객체글 사이의 참조 구조를 이용한 네비게이션 기반 정보접근 가능|
|스키마 구조의 유사|프로그램 내의 정보 구조와 데이터베이스 스키마 구조가 거의 유사함|


# 10. 데이터 독립성

## 3단계 스키마 구조의 정의

* 데이터베이스를 보는 관점(View)에 따라 사용자 관점, 통합모델 관점, 물리적 저장구조의 세가지 관점에서 아키텍쳐 및 이들간의 상호연관 관계를 정의한 데이터베이스 구조

|항목|설명|핵심|
|---|---|---|
|외부 스키마<br> (External Schema)|- 각 사용자나 응용프로그래머가 접근하는 데이터메이스<br> - 하나의 데이터베이스에 여러 개의 외부 스키마 존재|사용자 관점|
|개념 스키마<br> (Conceptual Schema)|- 모든 응용시스템, 사용자가 필요한 데이터를 통합한 조직 전체 데이터베이스<br> - 모든 개체, 제약조건, 접근권한, 보안, 무결성 규칙 포함|통합 관점|
|내부 스키마<br> (Internal Schema)|- 물리적 장치에서 Data가 실제 저장되는 방법을 표현<br> - 내부 레코드 형식, index유무, 내부 레코드 순서|물리적 저장구조|

## 데이터 독립성

* 데이터베이스의 논리적/물리적 구조를 변경하더라도, 응용프로그램이 영향을 받지 않는 특성

## 데이터독립성 종류

|독립성|설먕|
|---|---|
|논리적 독립성|데이터베이스 논리적 구조 <br> (개념스키마를 변경하더라도 기존 응용프로그램에 영향을 주지 않는 특성)|
|물리적 독립성|응용프로그램이나 데이터베이스 논리적 구조에 영향을 주지 않고, <br> DBMS가 데이터의 물리적 구조를 변경할 수 있는 특성|


## 사상(Mapping)

* 데이터 독립성을 지원하기 위한 데이터베이스 구조간의 사상(mapping)필요

|사상|내용|예|
|---|---|---|
|외부적/개념적사상<br>(논리적사상)|외부적 뷰와 개념적 뷰의 상호관련성 정의|사용하가 접근하는 형식에 따라 다른 타입의 필드를 가질수 있음. <br> 개념적 뷰의 필드 타입은 변화가 없음|
|개념적/내부적사상<br>(물리적사상)|개념적 뷰와 저장된 데이터베이스의 상호연관성 정의|만약 저장된 데이터베이스 구조가 바뀐다면 개념적/내부적 사상이 바뀌어야함, 그래야 개념적 스키마가 그대로 남아있게 됨|

# 11. 3단계 스키마의 뷰

## 뷰의 개념 및 장점

* 특정 테이블로부터 조건에 맞는 내용들을 추출하여 생성한 가상 테이블(물리적 미존재)
* 뷰 생성시에 정의한 뷰의 스키마 내용만이 저장되어 있다가 실행 시에 기본 테이블(base table)로부터 내용을 가져와 생성

|기능|상세설명|
|---|---|
|단순화|- 두 개 이상의 테이블을 조인하여 테이블을 형성, 검색, 이용, 갱신|
|다양한관점|- 통합된 데이터들을 하나의 형식으로 저장 <br> - 사용자에게 서로 다른 데이터 이름 밒 형식으로 데이터 제공|
|독립성|- 테이블의 구조가 변경되더라도 뷰의 구조는 변경할 필요없음 <br> - 데이터의 논리적 독립성 제공|
|데이터의 보안|- 데이터에 대한 접근을 제한|

## 뷰의 단점

* 한번 정의되면 변경할 수 없다. (ALTER 명령어 사용못함)
* 뷰는 삽입, 삭제, 변경에 대한 연산에 많은 제약 사항이 따른다.

# 12. 인덱스(Index)

## 정의

* 어떤 종류의 검색 연산을 최적화 하기 위해 탐색키 값과 해당 데이터의 위치 정보(Pointer)로 구성되어있는 데이터 구조
* 어떤 파일의 레코드들에 대한 효율적인 접근을 위해 레코드값과 주소를 쌍으로 관리하는 보조 데이터 구조

## 설계 절차

|분류|설명|
|---|---|
|접근 경로 수집|- 프로그램 설계서, 화면 설계서, 프로그램 처리 조건등 고려하여 예상되는 접근 경로 수집|
|분포도 조사로 후보칼럼 선정|- 분포도 = 데이터별 평균 로우수/테이블 총 로우수 * 100 <br> - 분포도가 10~15% 정도면 인덱스 칼럼 후보로 선정 <br> - 단일 컬럼으로 만족하지 않는 경우 조합컬럼의 분포도 조사 실시|
|접근 경로 결정|- 인덱스 후보 목록을 이용해 접근 유형에 따라 결정|
|컬럼 조합 및 순서결정|- 단일 컬럼의 분포도가 양호하면 단일 컬럼 인덱스로 확정 <br> - 하나 이상의 컬럼 조합 필요시 컬럼을 선행 선정 <br> - ORDER BY, GROUP BY 순서를 적용|

# 13. 관계형 데이터베이스의 릴레이션과 속성

## 릴레이션

* 데이터를 표의 형태로 표현, 구조를 나타내는 릴레이션 스키마와 실제 값들인 릴레이션 인스턴스로 구성

## 튜플

* 릴레이션을 구성하는 각각의 행을 말함, 속성의 모임으로 구성
* 파일 구조에서 레코드와 같은 의미
* 튜플의 수를 카디널리티(Cardinality)또는 기수, 대응수라고 부름

## 속성(Attribute)

* 속성은 데이터베이스를 구성하는 가장 작은 논리적 단위이며 개체의 특성을 기술
* 파일 구조상의 데이터 항목 또는 데이터 필드에 해당
* 속성의 수를 디그리(Degree)또는 차수라고 함

## 도메인

* 도메인 하나의 속성이 취할 수 있는 같은 타입의 원자값들의 집합
* 실제 속성 값이 나타날 때 그 값의 합법 여부를 시스템이 검사하는데에도 이용

## 릴레이션 특징

* 한 릴레이션에는 똑같은 튜플이 포함될 수 없음
* 한 릴레이션에 포함된 튜플 사이에는 순서가 없음
* 튜플들의 삽입, 삭제 등의 작업으로 인해 릴레이션은 시간에 따라 변함
* 릴레이션 스키마를 구성하는 속성들 간의 순서는 중요하지 않음
* 속성의 유일한 식별을 위해 속성의 명칭은 유일해야 함. 하지만, 속성을 구성하는 값은 동일한 값이 있을 수 있다.
* 릴레이션을 구성하는 튜플을 유일하게 식별하기 위해 속성들의 부분집합을 키로 설정
* 속성의 값은 논리적으로 더 이상 쪼개질 수 없는 원자값만을 저장

# 17. 물리적 모델링

## 정의

* 논리 데이터 모델을 DBMS의 특성을 고려하여 데이터베이스 저장 구조로 변환하는 과정

* 관계 변환(1:1관계, 순환관계, 배타적 관계)
* 속성 -> column 변환
* Primary UID -> 기본키(Primary Key) 변환

## 데이터베이스 물리적 설계 주요 Task

|---|---|
|저장공간 설계|- 상용 DBMS는 데이터를 저장하는 다양한 방식을 제공하므로, 업무에 적합한 저장공간에 대한 설계가 필요|
|무결성 설계|- 데이터의 정확성, 일관성, 유효성, 신뢰성을 위해 잘못된 데이터 변경으로부터 데이터를 보호하기 위해 무결성 설계가 필요|

|분산 설계|- 네트워크 상에서 논리적으로 연관된 데이터베이스가 물리적으로 분산되어 운영되는 것을 분산 데이터베이스라고 함 <br> - 분산 데이터베이스 설계를 통해 여러 개의 물리적인 데이터베이스를 논리적으로 단일 데이터베이스로 인식하기 위한 투명성을 필요로 함|
|보안 설계|- 데이터베이스 내의 정보가 비인가자에 의해 노출, 변조, 파괴되는 것을 막고 사용자가 원하는 작업을 하기 위해 필요한 자원에 대한 접근 제어를 설계하는 것이 필요|

# 18. 데이터베이스 키

## 정의

* 여러 개의 집합체를 담고 있는 하나의 엔티티 타입에서 각각의 엔터티를 구분할 수 있는 결정자
* 릴레이션 튜플을 유일하게 식별할 수 있는 속성의 집합

## 특징

* 유일성: 키 값으로 하나의 튜플을 유일하게 식별
* 최소성: 키가 둘 이상의 속성으로 구성되었을 경우 하나의 속성을 제거하면 유일성 파괴, 키는 모든 튜플을 유일하게 식별할 수 있는 최소의 속성들로 구성

## 데이터베이스 키의 종류

|구분|내용|예|
|---|---|---|
|수퍼키(Super Key)|- 유일성을 만족하는 집합, 최소성 만족 못함|<이름, 학과, 학년>|
|후보키(Candidate Key)|- 유일성과 최소성 만족하는 키|<학번>, <이름, 학과>|
|대체키(Alternative Key)|- 유일성과 최소성을 만족하는 키 중 기본키 이외의 속성|<학번>이 기본키일때, <이름, 학과>가 후보키|
|기본키(priamry Key)|- 여러 개의 후보키 중 하나 선정(최소 단위 구성)|<학번>|
|외래키(Foreign Key)|- 어느 한 릴레이션 속성의 집합이 다른 릴레이션 기본키로 이용되는 키|<학과코드>가 <이름, 학과코드(FK), 학년>으로 적용|
