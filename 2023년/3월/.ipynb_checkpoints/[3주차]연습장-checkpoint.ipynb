```python
def getWeatherInfo(fileName):
    now = datetime.now()
    nowDate = now.strftime('%Y-%m-%d %H:%M:%S')
    print(nowDate)
    
    if DEBUG == True:
        print(nowDate)
    
    base_url = 'https://search.naver.com/search.naver?sm=top_hty&fbm=0&ie=utf8&query='
    local_name = urllib.parse.quote('서울날씨')
    webpage = urllib.request.urlopen(base_url + local_name)
    soup = BeautifulSoup(webpage, 'html.parser')
    
    temps = soup.find('div', {'class': 'temperature_text'}).text
    temps = float(re.sub(r'[^0-9.]', '', temps))
    
    if DEBUG == True:
        print(temps)
        
    temp_info = soup.find('div', {'class': 'temperature_info'})
    temp_info = temp_info.findAll('dd', {'class': 'desc'})
    hum = temp_info[1].text
    if DEBUG == True:
        print(hum)
    
    fineDust = soup.find('li', {'class':'item_today level2'})
    fineDust = fineDust.find('span', {'class':'txt'}).text
    
    if DEBUG == True:
        print(fineDust)
        
    ultraFineDust = soup.find('li', {'class': 'item_today level2'})
    ultraFineDust = ultraFineDust.find('span', {'class': 'txt'}).text
    
    if DEBUG == True:
        print(ultraFineDust)
        
    f = open(fileName, 'a', newline = '', encoding = 'euc-kr')
    wr = csv.writer(f)
    weatherInfo = [nowDate, temps, hum, fineDust, ultraFineDust]
    wr.writerow(weatherInfo)
    print('write Done!')
    f.close()
```

```python
# 그리니치 천문대 표준시
now = datetime.now()
print(now)

fileName = now.strftime('%Y%m%d_%H%M.csv')
f = open(fileName, 'w', newline = '', encoding = 'euc-kr')
wr = csv.writer(f)
wr.writerow(['datetime', 'temp', 'hum', 'fineDust', 'ultraFineDust'])
f.close()

job1 = schedule.every(10).seconds.do(getWeatherInfo, fileName)

while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
def fileReader(filename):
    with open(filename, 'rt', encoding = 'utf-8') as f:
        text = f.readlines()
    return text

def fileWriter(filename, newText):
    with open(filename, 'wt', encoding = 'utf-8') as f:
        for line in newText:
            f.write(line + '\n')
            
def replaceStr(text, myDict):
    newText = []
    i = 0
    while True:
        if i < 3:
            # 불러온 한 줄의 문자열에 줄바꿈이 있으면 삭제
            text[i] = text[i].replace('\n', '')
        else:
            break
        # 하나의 문자열을 공백 기준으로 나눠주는 함수
        words = text[i].split()
        
        # 마지막 줄이 아니라면
        if i < 3:
            # myDict의 key와 value를 참조하여 단어 대체
            for key, value in myDict.items():
                if key == words[1]:
                    # 문자열을 다른 단어로 대체
                    temp = text[i].replace(key, value)
                    newText.append(temp)
        else:
            newText.append(text)
            
        i = i + 1
    return newText
```

```python
text = fileReader(readFile)
myDict = {'첫번째':'1번째', '두번째':'2번째', '세번째':'3번째'}
newText = replaceStr(text, myDict)
print(newText)
fileWriter(writeFile, newText)
```

```python
def replaceStr2(text, myDict):
    newText = []
    for myStr in text:
        # news.txt에서 읽어온 문자열은 불필요한 문자열 공백, 줄바꿈 삭제
        myStr = myStr.strip().replace('\n', '')
        # myDict의 key와 value를 참조하여 단어 대체하기
        # myStr을 dict에서 가지고 있는 다른 키값으로 모두 검색해서 변경하려면
        # 문자열 변경이 누적되어야 한다
        for key, value in myDict.items():
            myStr = myStr.replace(key, value)
        newText.append(myStr)
        
    return newText
```

```python
# 선형함수 증명
def myFunc(x):
    y = 2*x
    return y
```

```python
def isLinear(x1,x2, k):
    cond1 = False
    cond2 = False
    
    result1 = myFunc(x1 + x2)
    result2 = myFunc(x1) + myFunc(x2)
    result3 = myFunc(k*x1)
    result4 = k*myFunc(x1)
    
    if result1 == result2:
        cond1 = True
    if result3 == result4:
        cond2 = True
    
    if cond1 == cond2:
        return True
    else:
        return False
    
import random
counter = 0
for i in range(100):
    x1 = random.randrange(1, 200)
    x2 = random.randrange(10, 300)
    k = random.randrange(1, 20)
    
    result = isLinear(x1, x2, k)
    if result == True:
        counter += 1
    else:
        print('함수는 비선형함수입니다')
        break
if counter == 100:
    print('함수는 선형함수 입니다')
```
