### 위도와 경도를 이용해거 지도상에 표시

```python
# geo함수는 '~시'를 인식하지 못하므로 위치컬럼의 시를 없애준다
geo_addr1 = solar_df['위치'].unique()
geo_addr = []
for addr in geo_addr1:
    geo_addr.append(addr.replace('시', ''))
geo_addr
```

```python
# 패키지 설치
!pip install geopy
```

```python
# 개발자 로그인을 통해 카카오맵, 네이버지도에서 좌표변환하는 것이 한글화가 잘되어있어 편리
# 그런 경우가 아니라면 geo모듈을 사용한다
# 가입 없이 주소 -> 좌표변환
from geopy.geocoders import Nominatim
# 비어있는 리스트에 주소, 위도, 경도 할당
unique_addr = []
unique_lat = []
unique_lng = []

def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    # address가 한글로 입력했을때 패키지에서 한글주소를 인식 못하는 경우
    geo = geolocoder.geocode(address)
    # 한글주소로 인식이 안되면
    if geo == None:
        print('geo is null')
        return 0, 0
    else:
        return geo.latitude, geo.longitude

for addr in geo_addr:
    lat, lng = geocoding(addr)

    if lat == 0 and lng == 0:
        continue
    unique_addr.append(addr)
    unique_lat.append(lat)
    unique_lng.append(lng)
    print(lat, lng)
```

```python
# 위의 결과에서 geo is null인 부분을 찾으면
# ['서귀포 안덕면', '서귀포 남원읍', '서귀포 도순동','제주 삼도2동']가 누락되어있음을 알 수 있다
# 누락된 좌표를 추가해준다
ko_addr = ['서귀포 안덕면', '서귀포 남원읍', '서귀포 도순동','제주 삼도2동']
eng_addr = ["Andeok-myeon, Seogwipo-si", 'Namwon-eup, Seogwipo-si', '29beon-gil, Seogwipo-si', '7, Gija-gil, Jeju-si']

index = 0
for addr in eng_addr:
    print(ko_addr[index])
    lat, lng = geocoding(addr)
    if lat == 0 or lng == 0:
        continue
    unique_addr.append(ko_addr[index])
    unique_lat.append(lat)
    unique_lng.append(lng)
    print(lat, lng)
    index += 1
```

```python
# 전체 좌표 데이터프레임을 만들어준다
df_unique = pd.DataFrame({'주소' : unique_addr,
                         '경도(x)': unique_lng,
                         '위도(y)': unique_lat})
df_unique
```

## 팰린드롬(회문)연습

```python
# 문자열의 가운데를 기준으로 생각한 구문
def is_palindrome(word):
    palin = True
    lenStr = len(word)
    for i in range(0, lenStr // 2):
        if word[i] == word[lenStr - (i + 1)]:
            continue
        else:
            palin = False
    return palin
print(is_palindrome('result'))
print(is_palindrome('토마토'))
print(is_palindrome('aabbcc'))
print(is_palindrome('kyayk'))

```

```python
# 다른 방법
def is_palindrome(word):
    return word == word[::-1]

print(is_palindrome('테스트'))
print(is_palindrome('kyayk'))
print(is_palindrome('토마토'))
```

### 파이참환경에서 모듈로 만들기

```python
# 함수를 확장 시켜본다
# csv파일을 만들어 놓고 팰린드롬 문자열인지 확인해보는 함수를 만든다
# 함수 실행 후 확인한다

# 디버그 모드인 경우 True
DEBUG = False

def is_palindrome(word):
    lenStr = len(word)
    palin = True
    
    for i in range(0, lenStr // 2):
        if DEBUG:
            print(word[i], word[lenStr - (i + 1)])
            
        if word[i] == word[lenStr - (i + 1)]:
            continue
        else:
            palin = False
            
    return palin

# 팰린드롬인지 확인해주는 문자 출력함수
def resultPrint(palin):
    if palin:
        print('팰린드롬 문자열이 맞다')
    else:
        print('팰린드롬 문자열이 아니다')

# 파일을 불러온다
import pandas as pd
csvFile = 'C:/Users/chunc/Desktop/취업캠프/python/python/palindrome.csv'
df = pd.read_csv(csvFile, encoding = 'euc-kr')
# print(df)
# for문을 사용하여 팰린드롬 확인
for text in df['text']:
    palin = is_palindrome(text)
    resultPrint(palin)
```

```python
# for문을 사용하지 않고 파일에 직접 값을 입력
df['result'] = df['text'].apply(is_palindrome)
print(df)
```

## 태양광발전소 모듈화 하기

```python
# 위의 팰린드롬 예제처럼 앞서서 복습한 태양광 데이터를 활용하여 모듈화를 한다
# 마찬가지로 파이참환경에서 실행했다
import pandas as pd
import numpy as np

# 파일을 객체에 저장
csvFile = 'C:/Users/chunc/Desktop/취업캠프/python/python/제주특별자치도_신재생에너지발전시설현황_20200120.csv'
```

```python
# 전처리함수 만들기
def preProc(csvFile):
    # 1. 파일 불러오기
    energy_df = pd.read_csv(csvFile, encoding = 'euc-kr')
    
    # 1.1 불필요한 컬럼삭제
    energy_df.drop(['허가일자', '구분', '데이터기준일자'], axis = 1, inplace = True)
    # print(energy_df.head())
    
    # 2. 원동력종류 == '태양광'만 선별
    solar_df = energy_df[energy_df['원동력종류'] == '태양광']
    # print(solar_df.info())
    
    # 3. '사업개시일' 데이터타입을 datetime으로 변경
    solar_df['사업개시일'] = pd.to_datetime(solar_df['사업개시일'])
    # print(solar_df.info())
    
    # 4. '사업개시일'기준으로 정렬 ascending = True
    solar_df.sort_values(by = '사업개시일', ascending = True, inplace = True)
    # print(solar_df.head())
    
    # 5. '사업개시일'을 기준으로 새롭게 정렬했으므로 인덱스 번호를 리셋해준다
    solar_df.reset_index(drop = True, inplace = True)
    
    # 6. '설치행정시'에서 '시'를 삭제하고 그 문자열과 '설치지역'문자열을 합치기
    solar_df['위치'] = solar_df['설치행정시'].str.replace('시', '') + ' ' + solar_df['설치지역']
    # print(solar_df['위치'])
    
    return solar_df


# 전처리 함수 불러오기
solar_df = preProc(csvFile)

# 위도 경도 추출을 위한 함수
from geopy.geocoders import Nominatim
def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    # address가 한글로 입력했을때 패키지에서 한글주소를 인식 못하는 경우
    geo = geolocoder.geocode(address)
    # 한글주소로 인식이 안되면
    if geo == None:
        return 0, 0
    else:
        return geo.latitude, geo.longitude
# 전역변수(Global variance)
# 비어있는 리스트에 주소, 위도, 경도 할당
# 로컬변수로 선언하고 리턴값으로 설정해도 상관없다
unique_addr = []
unique_lat = []
unique_lng = []
# 위도와 경도 변환 함수 만들기
def addrToGeoPos(addr):
    for addrStr in addr:
        lat, lng = geocoding(addrStr)
        if lat == 0:
            print('geo is Null')
            continue
        # 전역변수의 요소를 추가
        unique_addr.append(addrStr)
        unique_lat.append(lat)
        unique_lng.append(lng)

        print(lat, lng)


unique_df = addrToGeoPos(solar_df['위치'].unique())

df_unique = pd.DataFrame({'주소': unique_addr,
                          '경도(x)': unique_lng,
                          '위도(y)': unique_lat})

print(df_unique)

# 3. 지도에 마커 출력하기
import pandas as pd
import folium

# 지도의 중심좌표, zoom_in을 얼마만큼 할지
korea_map = folium.Map(location=[33.38, 126.55], zoom_start=11)

print(df_unique)

for i in range(len(df_unique)):
    folium.Marker([df_unique.iloc[i][2], df_unique.iloc[i][1]], popup=df_unique.index[i]).add_to(korea_map)

korea_map.save('korea.html')
```

* str.replace(): 일부 단어라도 일치하는 단어를 찾아 바꾸는 경우
