```python
import pandas as pd
csvFile = 'C:/Users/chunc/Desktop/취업캠프/python/python/korean-idol.csv'
```

```python
df = pd.read_csv(csvFile)
df
```

```python
# MySQL에서는 한글컬럼명이 오류가 나므로, 컬럼명을 영문으로 저장
df.columns = ['name', 'group_name', 'company', 'gender', 'birth', 'height', 'ABO', 'brand']
```

```python
# MySQL에서 primary key로 사용할 컬럼은 고유한 이름을 가지고 있어야하므로
# 컬럼 데이터중에 primary key로 사용할 만한 컬럼이 없다면 index를 추가로 사용한다.

# csv저장시 index = True이므로 index도 csv에 자동저장, 단 index의 컬럼명은 저장되지 않으므로
# 인덱스 컬럼추가와 인코딩방식을 'utf-8'로 설정해주고 csv파일로 저장한다.
df.to_csv('myidol.csv', index_label = 'index', encoding = 'utf-8')
```
