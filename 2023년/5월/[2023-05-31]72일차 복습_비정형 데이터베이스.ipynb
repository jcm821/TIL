# 1. 빅데이터

## 정의

* 데이터베이스 관리도구의 능력을 넘어서는 대량(수십테라바이트)의 정형 또는 심지어 데이터베이스 형태가 아닌 비정형의 데이터 집합 조차 포함한 데이터로부터 가치를 추출하고 결과를 분석하는 기술 빅데이터의 필요성
* 아래 사진에서 다양성과 속도의 내용이 바뀌어져 있다.

## 정형 데이터

* 관계형 데이터베이스 시스템의 테이블과 같이 고정된 컬럼에 저장되는 데이터와 파일, 그리고 지정된 행과 열에 의해 데이터의 속성이 구별되는 스프레드시트 형태의 데이터도 있을 수 있음
* 관계형 데이터베이스 시스템의 정형 데이터를 비정형 데이터(Unstructured Data)와 비교할 때 가장 큰 차이점은 데이터의 스키마를 지원하는 것

## 반정형 데이터

* 데이터의 스키마 정보를 관리하는 DBMS와 데이터 내용이 저장되는 데이터 저장소로 구분되지만, 반정형 데이터는 데이터 내부에 정형 데이터의 스키마에 해당되는 메타데이터를 갖고 있으며, 일반적으로 파일 형태로 저장

## 비정형 데이터

* 데이터 세트가 아닌 하나의 데이터가 수집 데이터로 객체화
* 언어 분석이 가능한 텍스트 데이터나 이미지, 동영상 같은 멀티미디어 데이터가 대표적인 비정형 데이터임
* 웹에 존재하는 데이터의 경우 html 형태로 존재하며 반정형 데이터로 구분할 수 있지만, 특정한 경우 텍스트 마이닝을 통해 데이터를 수집하는 경우도 존재하므로 명확한 구분은 어려움

# 2. NoSQL

## 정의

* 관계형 DB의 한계를 벗어나, Web의 비정형 초고용량 데이터 처리를 위해 데이터의 읽기보다 쓰기에 중점을 둔, 수평적 확장이 가능하며 다수 서버들에 데이터 복제 및 분산 저장이 가능한 DBMS
* 초고용량 데이터 처리 등 성능에 특화된 목적을 위해, 비관계형 데이터 저장소에 비구조적인 데이터를 저장하기 위한 분산저장시스템

# 3. CAP이론

## CAP Theorem 정의

* 대용량 분산 데이터 저장소는 데이터 일관성(Consistency), 가용성(Availability), 단절내성(Partition Tolerance)을 모두 만족시키는 것이 불가능하므로 두 가지만 전략적으로 선택해야 한다는 이론

|특성|설명|
|---|---|
|Consistency|- 모든 노드들은 같은 시간에 같은 데이터를 보여줘야 함<br>(각각의 사용자가 항상 동일한 데이터를 조회)|
|Availability|- 몇몇 노드가 다운되어도 다른 노드들에게 영향을 주지 않아야 함|
|Partition Tolerance|- 일부 메시지를 손실하더라도 시스템은 정상 동작을 해야함<br>(물리적 네트워크 분산 환경에서 시스템이 잘 동작함)|
|C + A|- 시스템이 죽더라도 메시지 손실은 방지하는 강한 신뢰형<br> - 트랜잭션이 필요한 경우 필수적, <br> - 일반 RDBMS|
|C + P|- 모든 노드가 함께 퍼포먼스를 내야하는 성능형 <br> - 구글의 BigTable, HyperTable, Hbase|
|A + P|- 비동기화된 스토어 작업에 필수적 <br> - Dynamo, Apache Cassandra, CouchDB, Oracle Coherence|

# 4. ACID와 BASE

### ACID의 정의
* 데이터베이스 트랜잭션이 안전하게 수행된다는 것을 보장하기 위한 성질

### BASE의 정의
* ACID와 대조적으로 가용성과 성능을 중시하는 특성을 가진 분산 시스템의 특성

## NoSQL의 BASE 속성

|구분|내용|
|---|---|
|Basically Available|- 가용성을 중시, Optimistic Locking 및 큐 사용|
|Soft-State|- 노드의 상태는 외부에서 전송된 정보를 통해 결정됨<br> - 분산 노드 간 업데이트는 데이터가 노드에 도달한 시점에 갱신|
|Eventually Consistent|- 일시적으로 비일괁적인 상태가 되어도 일정시간 후에는 일관성이 있는 상태가 되는 성질|


## RDBMS의 ACID속성

|구분|내용|
|---|---|
|Atomicity<br>(원자성)|- 트랜잭션과 관련된 작업들이 모두 수행되었는지 모두 실행되지 않았는지를 보장하는 것|
|Consistency<br>(일관성)|- 트랜잭션이 실행을 성공적으로 완료하면 언제나 일관성있는 데이터베이스 상태로 유지하는 것|
|Isolation<br>(고립성)|- 트랜잭션 수행 시 다른 트랜잭션의 연산 작업이 끼어들지 못하도록 보장하는 것|
|Durability<br>(지속성)|- 성공적으로 수행된 트랜잭션은 영원히 반영되어야 하는 것|

## 정의

|특징|설명|기술|
|---|---|---|
|원자성(Atomicity)|분해가 불가능한 수행단위로 완전히 수행되거나 수행되지 않는 속성| Commit, rollback 2PC|
|일관성(Consistancy)|트랜잭션 성공, 실패에 따른 데이터 상태의 일관된 모습을 제공하는 속성|2PL, timestamp|
|고립성(Isolation)|트랜잭션 실행 중 중간연산 결과는 다른 트랜잭션에서 접근이 불가한 속성|Isolation Level|
|영속성(Durability)|성공적으로 완료된 트랜잭션은 어떤 고장에도 손실되진 않는 속성|Backup 회복|

# 7. 트랜잭션 보장 기법

## Locking 기법
* 트랜잭션이 데이터에 대하여 잠금(Lock)을 설정하면, 잠금을 설정한 트랜잭션이 해제(Unlock)할 때까지 데이터를 독점적으로 사용할 수 있는 기법

|종류|주요개념|
|---|---|
|공유Lock|- Lock한 트랜잭션은 데이터 항목에 대해 읽기만 가능 <br> - 다른 트랜잭션도 읽기만을 실행 할 수 있는 형태|
|전용Lock|- Lock한 트랜잭션은 데이터 항목에 대해서 읽기 쓰기 가능 <br> - 다른 트랜잭션은 읽기와 기록 모두 못하는 상태|


### Locking 규약
* 트랜잭션 T가 read(x), write(x)를 수행하기 전에 반드시 lock(x)연산을 실행
* 트랜잭션이 모든 실행을 종료하기 전에 반드시 unlock 실행

## 2PL(Tow Phase Locking) 기법
* 트랜잭션이 lock규약을 따르는 것만으로 직렬가능성을 보장하지 못하는 현상을 보완하는 기법

## Timestamp 기법
* 발생하는 트랜잭션에 timestamp tocken을 부여하고 timestamp 순서대로 실행하여 직렬 스케줄의 결과와 같게 보장하는 방법

## 낙관적 병행제어
* 트랜잭션이 실행되는 동안에는 아무런 검사를 실시하지 않고, 실행이 끝난 후 확인 단계에서 직렬가능성이 위배되는지 검사
* 갱신(update)보다는 read가 많은 트랜잭션에서 이점

# 9. 데이터베이스 회복

## 정의

* 데이터베이스에 장애가 발생했을 때 데이터베이스를 장애발생 이전의 내용에 모순이 없는 일관된 상태로 복원시키는 행위

## Redo

* 데이터베이스 내용자체가 손상된 경우 적용
* 가장 최근의 복제본을 적재시킨뒤 이 복제본 이후에 일어난 변경만을 로그를 이용하여 재실행

## Undo

* 데이터베이스 내용자체는 손상되지 않았지만 갱신하였거나 갱신중인 내용에 대한 신뢰성을 상실한 경우에 사용

## 로그기반 자연갱신 회복기법

* 트랜잭션이 부분완료 상태에 이르기 전까지 발생한 모든 변경 내용을 로그파일에 기록하고, 데이터베이스에 저장하는 것을 지연시키는 방법
* 부분완료 상태에 도달한 경우 로그파일을 참조하여 변경된 내용을 데이터베이스에 반영
* 트랜잭션 수행 도중 실패한 경우 로그파일의 내용만을 폐기하면 되므로 Undo 필요 없음

## 로그기반 즉시갱신 회복기법

* 트랜잭션 수행 과정의 변경내용을 데이터베이스와 로그파일에 동시에 저장하는 방법
* 트랜잭션 수행 도중 실패한 경우 로그파일의 내용을 참조하여 변경 이전의 값으로 수정하는 Undo과정을 거쳐야 함

## 검사점 회복기법

* 트랜잭션 수행 도중 발생하는 변경내용을 로그파일에 기록
* 일정시간 단위마다 체크포인트 발생
* 그때까지의 변경내용을 디스크에 저장
* 로그파일에 체크포인트 기록
* 장애 발생시 체크포인트 이전에 처리된 트랜잭션들은 회복작업에서 제외
* 검사시점 이후에 처리된 트랜잭션에 대해서만 회복작업(Redo or Undo)를 수행

## 그림자페이지 회복기법

* 트랜잭션이 실행되는 동안 현재 페이지 테이블과 그림자 테이블 이용
* 현재 페이지 테이블은 메모리에, 그림자페이지 테이블은 디스크에 저장
* 데이터베이스 트랜잭션의 시작지점에 현재 페이지 테이블의 내용과 동일한 그림자 페이지 테이블을 생성
* 트랜잭션의 변경 연산이 수행되면 현재 페이지 테이블의 내용만 변경하고 그림자 페이지 테이블의 내용은 변경하지 않음
* 트랜잭션이 성공적으로 완료될 경우, 현재 페이지 테이블의 내용을 그림자 페이지 테이블의 내용으로 저장
