# 캔틸레버(외팔보 구조, Cantilever)

참고 블로그: https://m.post.naver.com/viewer/postView.naver?volumeNo=10751817&memberNo=38387794

## 정의

* 한쪽 끝이 고정되고 다른 끝은 받쳐지지 않은 상태로 되어 있는 보로, 외팔보라고도 한다.
* 한쪽 끝은 고정되고 다른쪽 끝은 자유로운 들보이다.
* 마치 떠있는 듯한 디자인을 건축하거나, 하천 유속에 영향이 적은 교량을 설계할 수 있다.

## 기술 및 주의점

* 외관은 경쾌하나 같은 길이의 보에 비해 4배의 휨 모멘트를 박아 변형되기 쉬우므로 강도설계에 주의를 요한다.

* 외팔보는 고정단에 발생하는 휨 모멘트와 전단력을 통해 하중을 지지한다.

* 외팔보의 어떤 부분에도 보의 상단은 잡아당겨지고 하단은 압축을 받는 상태로 되어 있다.<br>
-> 철근 콘크리트의 외팔보에서는 보의 상단에 반드시 철근 배치

* 외팔보에 생기는 힘은 보의 선단에서 밑으로 향할수록 강해지기 때문에 보의 단면을 밑에서 크게 하고 선단부에서는 적게 하는 경우가 많다.

## 용도

* 외팔보는 교량, 탑, 등의 고정 구조물 외에도 항공기 날개 등에 대표적으로 사용된다.
* 주로 건물의 처마 끝, 현관의 차양, 발코니 등에 많이 사용됨
* 공중에 떠있는 듯한 동적인 공간 효과와 햇빛을 막거나 경쾌하게 보이기 위하여 근대 건축 뿐만 아니라 의자 디자인에 채용함으로써 종래의 의자와는 전혀 다른 새로운 구조를 탄생시켰다고 전해진다.
* 응력에 많이 견딜 수 있는 구조. 즉, 철골구조, 철근 콘크리트 구조 등을 쓰는 현대 구조법에 많이 쓰이는데 캔틸레버식 교량, 공장건축, 상점 건축은 물론 주택(차양)에도 많이 쓰인다. 석조에서는 매우 육중한 벽을 형성하게 되고 재료가 많이 든다.
