```python
# 람다함수를 이용
# 람다함수의 목적은 짧은 코드로 바로 실행할 수 있는 함수
```

```python
# 예시
mul_func = lambda x,y: x*y
```

```python
# x = df['Name'][0]
conv_name = lambda x:x.split(',')[1].split('.')[0]
# map함수는 series값을 받아와서 conv_name이라는 lambda함수에 하나씩 데이터를 전달하면서 전체 컬럼데이터를 실행하는 역할을 한다
df['Title'] = df['Name'].map(conv_name)
```

```python
df
```

```python
# 다른 방법
def myMul(x):
    return x.split(',')[1].split('.')[0]

df['Title'] = df['Name'].map(myMul)
df
```
