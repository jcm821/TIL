* agg()메서드: 사용자에게 다중집계작업을 간결하게 나타낼수 있게 한다.

agg()와 aggregate()는 동일한 메서드이며 사용방식은 아래와 같다.

result = object.agg(func, axis = 0, args, **kwargs)<br>
result = object.aggregate(func, axis = 0, args, **kwargs)

* 하나의 함수를 사용하는 것은 apply()메서드를 사용하는 것과 동일하다.
* 사용자는 문자열로 이름이 명명된 메서드를 입력할 수 있다.
* 이것들은 집약된 출력으로 구성된 Series로 반환한다.

```python
# 승차객 best(5)
# agg함수 활용
df.groupby('역명').agg({'승차총승객수':'sum'})

df1 = df.groupby('역명').agg({'승차총승객수':'sum'})
df1.sort_values(by = '승차총승객수', ascending = False).head()
```

```python
# 하차객 best(5)
df.groupby('역명').agg({'하차총승객수':'sum'})
df1 = df.groupby('역명').agg({'하차총승객수':'sum'})
df1.sort_values(by = '하차총승객수', ascending = False).head()
```

```python
# 승하차객 best(5)
df['승하차총승객수'] = df['승차총승객수'] + df['하차총승객수']

df.groupby('역명').agg({'승하차총승객수':'sum'})

df1 = df.groupby('역명').agg({'승하차총승객수':'sum'})
df1.sort_values(by = '승하차총승객수', ascending = False).head()
```
