### 2. 컴퓨터 및 주변기기 카테고리 내 세부항목들의 검색 성비를 확인

- 1번에서 예상했던 블랙프라이데이 이벤트와 연관이 있는지 각 세부 품목들에 대한 검색량 성비를 확인해봤다.
- 각 품목마다 22년10월1일부터 12월 31일 까지의 검색 성비를 보게 되면 컴퓨터/가전 카테고리 내에 어떤 품목이 여성의 검색량이 급증했는지 알 수 있을것으로 예상됨
- 성비그래프는 꺾은선 형태의 그래프로 도출했음(특정 날짜 사이를 비교하는 것이라 추세를 파악하기 용이할 것으로 생각해서)
- 왼쪽 눈금레이블은 여성기준, 오른쪽 눈금레이블은 남성기준이다.(태블로에서는 그래프에 커서를 대면 레이블 상관없이 비율이 나오나, 사진상에서는 혼동이 올 수 있으므로)
- 만약 다른 형태로 변경하고자 한다면 적극 수용
