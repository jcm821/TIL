# 통신사 데이터 지도히트맵

* 통신사 데이터로는 자치구별로 총인구수, 1인 가구수, 쇼핑 서비스 사용 미추정 인구, 쇼핑 서비스 사용일수 등의 비율 분포를 보고자 했음
* 1차 EDA 시각화 시에 쇼핑서비스 사용일수만 평균을 사용하고 나머지 항목들에 대해선 합계로 산출했음
* 히트맵의 색상 및 수치 수정은 추후 필요할 시 진행

## 자치구별 총인구수 지도 히트맵

* 송파구, 강서구, 노원구 순으로 총인구수가 많음이 보임
* 학교 및 학원과 같은 교육환경이 밀집되어있는 자치구에 인구수가 많이 분포되어 있음
* 대조적으로 기업이 몰려있는 중구, 종로구, 용산구, 금천구 등은 상대적으로 총인구수 분포가 낮음을 알 수 있음

## 자치구별 1인가구수 지도 히트맵

* 관악구, 강서구, 강남구 순으로 1인 가구수의 분포가 많음
* 눈에 띄게 관악구의 1인가구수 비율이 높음
* 독거노인, 고시원원룸, 자취 등 타 자치구보다 저렴한 집값(부동산적인 요소)의 영향이 있을것으로 예상됨

## 자치구별 쇼핑 서비스 사용 미추정 인구수 지도 히트맵

* 관악구, 노원구, 강서구 순으로 쇼핑 서비스 사용 미추정 인구수 분포도가 많음
* 관악구, 강서구의 경우 앞서 1인가구수 분포에서도 상위권이었던 것을 생각하면 사용 미추정 인구수 대부분이 서비스에 익숙하지 않은 연령대일 것으로 예상됨

## 자치구별 평균 쇼핑 서비스 사용일수 지도 히트맵

* 은평구, 노원구, 도봉구 순으로 평균 쇼핑서비스 사용일수가 높음
* 상대적으로 강북에 위치한 자치구에서 비율이 높음 -> 동대문 종로구 등 상업단지가 밀집되어 있는 자치구에 접근이 용이하다 판단됨

# 택배데이터

* 샘플데이터이므로 데이터의 수가 많지 않아 세부적인 인사이트를 얻기엔 부족하다 판단하여 자치구별 대분류 카테고리의 비중만을 파악
