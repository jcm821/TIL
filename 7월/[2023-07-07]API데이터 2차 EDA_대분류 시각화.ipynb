## 결과

### 20대

- 20대 남성: 디지털가전 > 생활/건강 > 패션잡화 > 패션의류
### 30대
- 30대 여성: 패션의류 > 생활/건강 > 출산/육아 > 패션잡화
- 30대 남성: 생활건강 > 디지털/가전 > 스포츠/레저 > 패션잡화
### 40대
- 40대 여성: 패션의류 > 생활/건강 > 패션잡화 > 출산/육아
- 40대 남성: 생활건강 = 디지털/가전 > 스포츠/레저 > 패션잡화
### 50대
- 50대 여성: 패션의류 > 생활/건강 > 패션잡화 > 식품
- 50대 남성: 생활건강 > 디지털/가전 > 스포츠/레저 > 패션잡화
### 60대
- 60대 여성: 패션의류 > 생활/건강 > 식품 > 패션잡화
- 60대 남성: 디지털/가전 = 스포츠/레저 > 생활건강 > 식품
