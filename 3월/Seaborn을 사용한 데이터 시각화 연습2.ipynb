# 샘플데이터를 이용한 데이터 시각화

```python
# 패키지 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# titanic데이터 불러오기
titanic = sns.load_dataset('titanic')
titanic.head()
```

## 데이터 정보

* survived: 생존여부
* pclass: 좌석등급
* sex: 성별
* age: 나이
* sibsp: 형제자매 + 배우자 숫자
* parch: 부모자식 숫자
* fare: 요금
* embarked: 탑승 항구
* class: 좌석등급 (영문)
* who: 사람 구분
* deck: 데크
* embark_town: 탑승 항구 (영문)
* alive: 생존여부 (영문)
* alone: 혼자인지 여부

```python
# tips데이터 불러오기
tips = sns.load_dataset('tips')
tips.head()
```

## 데이터 정보

* total_bill: 총 합계 요금표
* tip: 팁
* sex: 성별
* smoker: 흡연자 여부
* day: 요일
* time: 식사 시간
* size: 식사 인원

## 1. CountPlot

```python
# 배경을 darkgrid로 설정
sns.set(style = 'darkgrid')
```

### 1-1. 세로로 그리기

```python
sns.countplot(x = 'class', hue = 'who', data = titanic)
plt.show()
```

### 1-2. 가로로 그리기

```python
sns.countplot(y = 'class', hue = 'who', data = titanic)
plt.show()
```

### 1-3. 색상 파레트 설정

```python
sns.countplot(x = 'class', hue = 'who', palette = 'copper', data = titanic)
plt.show()
```

## 2. distplot

* matplotlib의 'hist'그래프와 'kdeplot'을 통합한 그래프
* 분포와 밀도를 확인할 수 있음

```python
# 샘플 데이터 생성
x = np.random.randn(1000)
x
```

### 2-1. 기본 distplot

```python
sns.distplot(x)
plt.show()
```

### 2-2. 데이터가 Series인 경우

```python
x = pd.Series(x, name = "x variable")
x
```

```python
sns.distplot(x)
plt.show()
```

### 2-3. rugplot

'rug'는 'rugplot'이라고도 불리우며, 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 데이터들의 위치 및 분포를 보여준다.

```python
sns.distplot(x, rug=True, hist=False, kde=True)
plt.show()
```

### 2-4. kde(kernel density)

kde는 histogram보다 부드러운 형태의 분포 곡선을 보여준다.

```python
sns.distplot(x, rug=False, hist=False, kde=True)
plt.show()
```

### 2-5. 가로로 표현하기

```python
sns.distplot(x, vertical = True)
```

### 2-6. 색상 바꾸기

```python
sns.distplot(x, color = 'g')
plt.show()
```

## 3. heatmap

* 색상으로 표현할 수 있는 다양한 정보를 일정한 이미지위에 열분포 형태의 비쥬얼한 그래픽으로 출력하는 것이 특징이다.

### 3-1. 기본 heatmap

```python
uniform_data = np.random.rand(10, 12)
sns.heatmap(uniform_data, annot=True)
plt.show()
```

### 3-2. pivot table을 활용하여 그리기

```python
tips
```

```python
pivot = tips.pivot_table(index = 'day', columns = 'size', values = 'tip')
pivot
```

```python
# pivot테이블을 토대로 히트맵을 그려본다.
sns.heatmap(pivot, cmap = 'Blues', annot = True)
plt.show()
```

### 3-3. correlation(상관관계)를 시각화

```python
# titanic 데이터 사용
titanic.corr()
```

```python
sns.heatmap(titanic.corr(), annot = True, cmap = "YlGnBu")
plt.show()
```

## 4. pairplot

* pairplot은 그리드(grid)형태로 각 집합의 조합에 대해 히스토그램과 분포도를 그려준다.
* 숫자형 column에 대해서만 그려준다.

```python
tips.head()
```

### 4-1.기본 pairplot그리기

```python
# tips 데이터
sns.pairplot(tips)
plt.show()
```

```python
# iris 데이터
iris = sns.load_dataset('iris')
sns.pairplot(iris)
plt.show()
```

### 4-2. hue 옵션으로 특성 구분

```python
sns.pairplot(tips, hue = 'size')
plt.show()
```

### 4-3. 컬러 파레트 적용

```python
sns.pairplot(tips, hue = 'size', palette = 'rainbow')
plt.show()
```

### 4-4. 사이즈 적용

```python
sns.pairplot(tips, hue = 'size', palette = 'rainbow', height = 5,)
plt.show()
```

## 5. lmplot

* column간의 선형관계를 확인하기에 용이한 차트
* outlier(이상치)도 같이 짐작할 수 있다.

### 5-1. 기본 lmport

```python
sns.lmplot(x = 'total_bill', y = 'tip', height = 8, data = tips)
plt.show()
```

### 5-2. hue 옵션으로 다중 선형관계 그리기

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", height=8, data = tips)
plt.show()
```

### 5-3 col옵션을 추가하여 그래프를 별도로 그릴 수 있다.

* 또한 'col_wrap'으로 한 줄에 표기할 column의 개수를 명시할 수 있다.

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='day', col_wrap=2, height=6, data=tips)
plt.show()
```

## 6.  relplot

* 두 column간 상관관계를 보지만, lmplot처럼 선형관계를 따로 그려주지는 않는다.

### 6-1. 기본 relplot

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'day', data = tips)
plt.show()
```

### 6-2. col 옵션으로 그래프 분할

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'day', col = 'time', data = tips)
plt.show()
```

### 6-3. row와 column에 표기할 데이터column 선택

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'day', row = 'sex', col = 'time', data = tips)
plt.show()
```

### 6-4. 컬러 파레트 적용

```python
sns.relplot(x = "total_bill", y = "tip", hue = "day", row = "sex", col = "time", palette = 'CMRmap_r', data = tips)
plt.show()
```
