# 빅데이터 분석기사 실기

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
# 수질 음용성 여부 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/train.csv')
df.head()
```

```python
# 93. ph값은 상당히 많은 결측치를 포함한다. 결측치를 제외한 나머지 데이터들 중 사분위값 기준 하위 25%의 값들의 평균값은?
target = df['ph'].dropna()
ans = target.loc[target <= target.quantile(0.25)].mean()
print(ans)
```

```python
# 의료 비용 예측 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv')
df.head()
```

```python
# 94. 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?
high = df.loc[df['smoker'] == 'yes']['charges'].quantile(0.9)
high2 = df.loc[df['smoker'] == 'no']['charges'].quantile(0.9)

mean_yes = df.loc[(df['smoker'] == 'yes') & (df['charges'] >= high)]['charges'].mean()
mean_no = df.loc[(df['smoker'] == 'no') & (df['charges'] >= high2)]['charges'].mean()
ans = mean_yes - mean_no
print(ans)
```

```python
# 킹카운터 주거지 가격예측문제 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv')
df.head()
```

```python
# 95. bedrooms 의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와 하위 10%값의 차이를 구하여라
answer = df.loc[df['bedrooms'] == df['bedrooms'].value_counts().index[0]]['price'].quantile(0.9) \
- df.loc[df['bedrooms'] == df['bedrooms'].value_counts().index[0]]['price'].quantile(0.1)
print(answer)
```

```python
# 대학원 입학가능성 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/train.csv')
df.head()
```

```python
# 96. Serial No. 컬럼을 제외하고 ‘Chance of Admit’을 종속변수, 나머지 변수를 독립변수라 할때,
# 랜덤포레스트를 통해 회귀 예측을 할 떄 변수중요도 값을 출력하라 (시드값에 따라 순서는 달라질수 있음)
from sklearn.ensemble import RandomForestRegressor

df_t = df.drop([df.columns[0]], axis = 1)
x = df_t.drop([df.columns[-1]], axis = 1)
y = df_t[df.columns[-1]]

ml = RandomForestRegressor()

ml.fit(x, y)

result = pd.DataFrame({'importance': ml.feature_importances_}, x.columns).sort_values('importance', ascending = False)
display(result)
```

```python
# 레드 와인 퀄리티 예측 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/train.csv')
df.head()
```

```python
# 97. quality 값이 3인 그룹과 8인 데이터그룹의 각 컬럼별 독립변수의 표준편차 값의 차이를 구할때 
# 그값이 가장 큰 컬럼명을 구하여라
ans = (df.loc[df['quality'] == 8].std() - df.loc[df['quality'] == 3].std()).sort_values().index[-1]
print(ans)
```

```python
# 약물 분류 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/train.csv')
df.head()
```

```python
# 98. 남성들의 연령대별 (10살씩 구분 0~9세 10~19세 …) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라
pre = df.loc[df['Sex'] == 'M']
pre2 = pre.copy()
pre2['Age2'] = pre['Age'] //10 * 10

ans = pre2.groupby('Age2')['Na_to_K'].mean().to_frame()
display(ans)
```

```python
# 사기회사 분류 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/audit/train.csv')
df.head()
```

```python
# 99. 데이터의 Risk 값에 따른 score_a와 score_b의 평균값을 구하여라
ans = df.groupby('Risk')[['Score_A', 'Score_B']].mean()
display(ans)
```

```python
# 센서데이터 동작유형 분류 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/train.csv')
df.head()
```

```python
# pose값에 따른 각 motion컬럼의 중간값의 가장 큰 차이를 보이는 motion컬럼은 어디이며 그값은?
t = df.groupby('pose').median().T
dfs = abs(t[0] - t[1]).sort_values().reset_index()
dfs[dfs[0] == dfs[0].max()]['index'].values
```

```python
# 현대 차량 가격 분류문제 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/train.csv')
df.head()
```

```python
# 101. 정보(row수)가 가장 많은 상위 3차종의 price값의 각 평균값은?
ans = df.loc[df['model'].isin(df['model'].value_counts().index[:3])].groupby('model')['price'].mean().to_frame()
display(ans)
```

```python
# 당뇨여부판단 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/train.csv')
df.head()
```

```python
# 102. Outcome 값에 따른 각 그룹의 각 컬럼의 평균 차이를 구하여라
ans = df.groupby('Outcome').mean().diff().iloc[1,:]
print(ans)
```

```python
# 넷플릭스 주식 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/nflx/NFLX.csv')
df.head()
```

```python
# 103. 매년 5월달의 open가격의 평균값을 데이터 프레임으로 표현하라
df['Date'] = pd.to_datetime(df['Date'])

target = df.groupby(df['Date'].dt.strftime('%Y-%m')).mean()
ans = target.loc[target.index.str.contains('-05')]['Open']
print(ans)
```

```python
# NBA 선수 능력치 데이터
pd.set_option('display.max_columns', 50)
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/nba/nba.csv', encoding = 'latin', sep = ';')
df.head()
```

```python
# 104. Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라
result = round(df[df['Tm'] == 'TOR']['Age'].mean(), 4)
print(result)
```

```python
# 105. Pos 컬럼은 포지션을 의미한다.
# 전체 선수 중 최소나이대의 선수들을 필터하고 그들 중 가장 많은 포지션은 무엇인지 확인하라
result = df[df['Age'] == df['Age'].min()]['Pos'].value_counts().index[0]
print(result)
```

```python
# 106. 선수들의 이름은 first_name+ 공백 + last_name으로 이루어져 있다.
# 가장 많은 first_name은 무엇이며 몇 회 발생하는지 확인하라
result = df['Player'].str.split().str[0].str.lower().value_counts().head(1)
print(result)
```

```python
# 107. PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라
result = df.groupby('Pos')['PTS'].mean().sort_values()
print(result)
```

```python
# 108. G컬럼은 참여한 경기의 숫자이다.
# 각 팀별로 가장 높은 경기참여 수를 가진 선수들의 경기 참여 숫자의 평균을 구하여라
result = df.sort_values(['Tm', 'G']).groupby('Tm').tail(1)['G'].mean()
print(result)
```

```python
# 109. Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?
result = df[(df['Tm'] == 'MIA') & (df['Pos'].isin(['C', 'PF']))]['MP'].mean()
print(result)
```

```python
# 110. 전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?
result = df[df['G'] >= df['G'].mean() * 1.5]['3P'].mean()
print(result)
```

```python
# 111. Age의 평균 이상인 그룹과 평균 미만인 그룹간의 G값의 평균의 차이는?
result = df[df['Age'] >= df['Age'].mean()]['G'].mean() - df[df['Age'] < df['Age'].mean()]['G'].mean()
print(result)
```

```python
# 112. 평균나이가 가장 젊은 팀은 어디인가
result = df.groupby('Tm')['Age'].mean().sort_values().index[0]
print(result)
```

```python
# 113. Pos그룹별 평균 MP값을 구하여라
result = df.groupby('Pos')['MP'].mean()
print(result)
```
