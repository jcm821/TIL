## 자동차 보험가입 예측데이터

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/train.csv')
df.head()
```

* Vehicle_Age 값이 2년 이상인 사람들만 필터링 하고 그중에서
Annual_Premium 값이 전체 데이터의 중간값 이상인 사람들을 찾고, 그들의 Vintage값의 평균을 구하여라

```python
# Vehicle_Age의 값 확인
df['Vehicle_Age'].unique()
```

```python
answer = df[(df['Vehicle_Age'] == '> 2 Years') & (df['Annual_Premium'] >= df['Annual_Premium'].median())]['Vintage'].mean()
print(answer)
```

* vehicle_age에 따른 각 성별(gender)그룹의 Annual_Premium값의 평균을 구하여 아래 테이블과 동일하게 구현하라

```python
mean_df = df.groupby(['Gender', 'Vehicle_Age'], as_index = False)['Annual_Premium'].mean()
mean_df.pivot(index = 'Vehicle_Age', columns = 'Gender', values = 'Annual_Premium')
```

## 핸드폰 가격 예측데이터

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv')
df.head()
```

* price_range 의 각 value를 그룹핑하여 각 그룹의 n_cores 의 빈도가 가장높은 value와 그 빈도수를 구하여라

```python
answer =df[['price_range','n_cores']].groupby(['price_range','n_cores']).size().sort_values().groupby(level=0).tail(1)
answer
```

* price_range 값이 3인 그룹에서 상관관계가 2번째로 높은 두 컬럼과 그 상관계수를 구하여라

```python
corrdf = df.loc[df['price_range'] == 3].corr().unstack().sort_values(ascending = False)
answer = corrdf.loc[corrdf != 1].reset_index().iloc[1]
print(answer)
```

## 비행탑승 경험 만족도 데이터

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv')
df.head()
```

* Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 ‘neutral or dissatisfied’ 보다 ‘satisfied’의 수가 더 높은 Class는 어디 인가?

```python
# 데이터 불러오
answer = df.loc[df['Arrival Delay in Minutes'].isnull()].groupby(['Class', 'satisfaction'], as_index = False).size()\
.pivot(index = 'Class', columns = 'satisfaction')
result = answer[answer['size']['neutral or dissatisfied'] < answer['size']['satisfied']]
result
```

## 수질 음용성 여부 데이터

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/train.csv')
df.head()
```

* ph값은 상당히 많은 결측치를 포함한다. 결측치를 제외한 나머지 데이터들 중 사분위값 기준 하위 25%의 값들의 평균값은?

```python
target = df['ph'].dropna()
answer = target.loc[target <= target.quantile(0.25)].mean()
print(answer)
```

## 의료 비용 예측 데이터

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv')
df.head()
```

* 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

```python
high = df.loc[df['smoker'] == 'yes']['charges'].quantile(0.9)
high2 = df.loc[df['smoker'] == 'no']['charges'].quantile(0.9)

mean_yes = df.loc[(df['smoker'] == 'yes') & (df['charges'] >= high)]['charges'].mean()
mean_no = df.loc[(df['smoker'] == 'no') & (df['charges'] >= high2)]['charges'].mean()
answer = mean_yes - mean_no
print(answer)
```

## 킹카운티 주거지 가격 예측문제 데이터

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice//train.csv')
df.head()
```

* bedrooms 의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와 하위 10%값의 차이를 구하여라

```python
answer = df.loc[df['bedrooms'] == df['bedrooms'].value_counts(ascending = False).index[0]]['price'].quantile(0.9) - \
df.loc[df['bedrooms'] == df['bedrooms'].value_counts(ascending = False).index[0]]['price'].quantile(0.1)
print(answer)
```

## 대학

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/train.csv')
df.head()
```

* Serial No. 컬럼을 제외하고 ‘Chance of Admit’을 종속변수, 나머지 변수를 독립변수라 할때, 랜덤포레스트를 통해 회귀 예측을 할 떄 변수중요도 값을  출력하라 (시드값에 따라 순서는 달라질수 있음)

```python
from sklearn.ensemble import RandomForestRegressor

df_t = df.drop([df.columns[0]], axis = 1)
x = df_t.drop([df.columns[-1]], axis = 1)
y = df_t[df.columns[-1]]

rf = RandomForestRegressor()

rf.fit(x, y)

result = pd.DataFrame({'importance': rf.feature_importances_}, x.columns).sort_values('importance', ascending = False)
display(result)
```

## 레드와인 퀄리티 예측 데이터

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/train.csv')
df.head()
```

* quality 값이 3인 그룹과 8인 데이터그룹의 각 컬럼별 독립변수의 표준편차 값의 차이를 구할때 그값이 가장 큰 컬럼명을 구하여라

```python
answer = (df.loc[df['quality'] == 8].std() - df.loc[df['quality'] == 3].std()).sort_values().index[-1]
print(answer)
```

## 약물 분류 데이터

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/train.csv')
df.head()
```

* 남성들의 연령대별 (10살씩 구분 0~9세 10~19세 …) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라

```python
pre = df.loc[df['Sex'] == 'M']
pre2 = pre.copy()
pre2['Age2'] = pre['Age'] // 10 * 10
answer = pre2.groupby('Age2')['Na_to_K'].mean().to_frame()
display(answer)
```
