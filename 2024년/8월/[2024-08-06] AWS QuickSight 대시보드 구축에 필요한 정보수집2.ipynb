* AWS 파이프라인의 구조중 가장 일반적으로 사용되는 구조중에 하나, 서버를 프로비저닝하거나 관리하지 않아도 코드를 실행할 수 있도록 해주는 컴퓨팅 서비스

2. **Batch Processing**: 통계 데이터 또는 다양한 테이블을 Join해서 인사이트를 얻고 싶다면 이 프로세싱이 적합
    - Ex. AWS Athena, AWS Glue, AWS Redshift

## 2. Data Lake: S3(Simple, Storage, Service)

* AWS(Amazon Web Service)에서 제공하는 인터넷 스토리지 서비스
* 데이터 베이스나 각종 로그의 데이터들이 통계 및 분석 툴로 가기 전, 데이터를 일관된 형태로 보관하는 역할

#### 장점
* 높은 내구도를 자랑하며 정보를 안전하게 저장 할 수 있다.
* 저렴한 비용으로 사용이 가능합니다(ec2에 이미지, 영상등을 저장하며 비용이 만만치 않다.)
* 보안성이 뛰어납니다 (SSL을 통하여 데이터 전송과 암호화를 하므로 해킹 걱정이 적다.)
* 속도가 빠릅니다 (각 지역에 맞게 선택하며, 업/다운로드 시 지역시간 최소화를 위한 멀티 파트 업로드를 지원)

### 데이터 마이그레이션 필요이유
* **데이터 마이그레이션**: 데이터를 한 시스템에서 다른 시스템으로 옮기는 과정
1. 기술적 진보 및 업그레이드: 기존의 데이터베이스나 저장 시스템이 더 이상 효율적이지 않거나 최신 기능을 지원하지 않을 경우, 최신 기술로 업그레이드하기 위해 데이터 마이그레이션이 필요
2. 비용 절감: 오래된 하드웨어나 시스템을 유지하는 데 드는 비용이 높아질 수 있음 -> 클라우드 서비스로 마이그레이션하면 인프라 관리 비용을 줄이고, 필요에 따라 자원을 유연하게 사용할 수 있어 비용 절감 효과 가능
3. 데이터 통합: 여러 데이터 소스에서 데이터를 하나의 통합된 시스템으로 이동하여 보다 일관되고 통합된 데이터 관리를 할 수 있음 -> 데이터 분석 및 비즈니스 인사이트를 향상시킬 수 있음
4. 성능 향상 및 확장성: 새로운 데이터베이스나 스토리지 솔루션은 더 나은 성능과 필요에 따라 확장할 수 있는 유연성을 제공 -> 대규모 데이터 처리나 애플리케이션 성능 최적화에 중요
5. 규정 준수 및 보안: 데이터 보안 표준이나 규정이 변경되었을 때, 이를 준수하기 위해 데이터 마이그레이션이 필요. 새로운 시스템으로의 마이그레이션을 통해 보안 강화와 규정 준수를 보장할 수 있음
6. 비즈니스 요구사항 변화: 비즈니스의 전략적 변화나 새로운 서비스 출시에 따라 데이터 구조나 접근 방식이 바뀔 수 있음 -> 기존 데이터를 새로운 요구사항에 맞게 조정하기 위해 마이그레이션이 필요

### GPT에 물어본 것
#### DMS를 사용해야 하는 경우
* 데이터베이스 간 마이그레이션이 필요한 경우, 예를 들어, 온프레미스 Oracle 데이터베이스를 AWS의 RDS로 마이그레이션하려는 경우.
* 실시간 데이터 동기화 또는 연속 복제가 필요한 경우.
* 데이터베이스의 전체 구조와 데이터를 빠르게 복제해야 할 때.
#### Glue를 사용해야 하는 경우
* 다양한 소스에서 데이터를 가져와 ETL 작업을 수행하고 데이터 레이크에 통합하고자 하는 경우.
* 데이터를 분석용으로 준비하기 위해 데이터 변환 작업이 필요한 경우.
* 데이터를 데이터 카탈로그에 등록하여 손쉽게 검색하고 사용하고자 하는 경우.
#### 활용방안
* 데이터베이스 간 마이그레이션이 주된 필요 사항이라면 DMS를 사용하는 것이 더 적합할 수 있습니다. 하지만, S3에 데이터를 저장하고 ETL 작업을 수행하며 데이터 분석 및 시각화까지의 전체 파이프라인을 구축하는 것이 목적이라면 Glue와 함께 Athena, QuickSight를 사용하는 접근 방식이 적절

## 4. AWS Glue(크롤러)

* 데이터의 스키마와 메타데이터를 관리
* 완전 관리형 ETL(추출, 변환 및 로드) 서비스, 간단하게 여러 데이터 스토어 및 스트림 간에 원하는 데이터를 분류, 정리, 보강, 이동할 수 있다.
* 서버리스이므로 설정하거나 관리할 인프라가 없다.
* Data Store의 경우, 크롤러를 정의하여 메타데이터 테이블 정의로 AWS Glue Data Catalog를 채운다.
* 크롤러를 데이터 스토어로 지정하면 크롤러는 데이터 카탈로그에 테이블 정의를 생성한다. 테이블 정의에 관련해 AWS Glue Data Catalog는 다른 필수 메타데이터를 포함하여 ETL 작업을 정의한다.
* 스크립트를 생성하여 데이터를 변환(Transform)할 수 있다 -> 스크립트를 AWS Glue 콘솔이나 API에 제공
* AWS Glue를 사용하여 데이터 웨어하우스 또는 데이터 레이크의 스토리지를 조직하고 정리하고 인증할 수 있다.

### ETL Job
* DMS와 마찬가지로 AWS Glue 또한 마이그레이션 기능을 제공
* AWS Glue에서 ETL job을 생성하려고 콘솔에 접속해보면, GUI환경 또는 스크립트 작성을 통해서 원하는 ETL/Migration Job을 생성할 수 있도록 도와준다.

### Data Catalog
* Raw데이터를 파싱해서 유의미한 형태의 Database로 만드는 기능
* RDS, 서비스 로그 등등 다양한 형태의 데이터/로그를 S3에 수집, 이 데이터들을 테이블 형태로 만들어줄 수 있는 것

### 주요 용어
#### data store
* 데이터를 지속적으로 저장하기 위한 저장소(Amazon S3 버킷 및 관계형 데이터베이스), 프로세스 또는 변환에 대한 입력(input)으로 사용되는 데이터 저장소
#### 크롤러(Crawle)
* 데이터 스토어(소스 또는 대상)에 연결하는 프로그램은 Classifier의 우선 순위 지정 목록을 통해 데이터의 스키마를 결정한 다음 AWS Glue Data Catalog에 메타데이터 테이블을 생성
#### Glue Data Catalog
* Glue의 영구적 메타데이터 스토어
테이블 정의, 작업 정의 및 기타 관리 정보를 포함하여 AWS Glue 환경을 관리
#### Job
* ETL 작업을 수행하는 데 필요한 비즈니스 로직
* 변환 스크립트(Trasform Script), data sources, and data targets으로 구성

## 4-1. DMS(Data Migration Service)

* DMS 활용 -> 다운타임 없이 마이그레이션이 가능하고 설정을 해두면, 스키마 변환도 간단히 구현이 가능
* 원하는 컬럼 또는 테이블만을 마이그레이션하는 것과 같은 다양한 맞춤 구성이 가능

### 복제 인스턴스 생성
* 마이그레이션 작업을 수행할 서버
* 특별한 목적이 있는 게 아니라면, '퍼블릭 엑세스 가능'은 꺼두는 것을 권장함

### 엔드 포인트 생성

* 마이그레이션의 원본 DB를 가리키는 '소스'엔드포인트와 옮겨질 목적지인 '대상'엔드포인트를 생성해줘야 한다.

#### 소스 엔드포인트
* 소스 엔드포인트는 RDS에 생성되어 있는 DB를 선택해주고, DMS가 사용할 DB 유저의 username과 password만 입력해주면 된다.
* 나중에 마이그레이션 Task를 수행할 때, DB User의 권한 에러가 발생할 수 있는데, Task의 로그를 확인해서 필요한 권한들을 추가해주어야한다.
* REPLICATION CLIENT, REPLICATION SLAVE 권한은 반드시 추가하는 것을 권장함

#### 대상 엔드포인트

* 대상 엔드포인트를 만들때는 IAM Role의 ARN을 추가해주는 부분을 적절한 IAM Role을 추가해주어야한다.
* 대상 엔드포인트에 사용할 IAM Role은 아래의 과정에 따라서 생성해주면 된다.

#### 주의사항

* 엔드포인트 설정에서 AddColumnName을 true로 설정해 주어야한다. 그렇지 않으면 컬럼명이 마이그레이션되지 않는다.
* DataFormat 옵션을 추가하고, parquet을 입력해주는 것이 좋다. 기본 값은 csv인데 csv형태로 마이그레이션을 할 경우에는 데이터에 ,가 있으면 제대로 마이그레이션이 되지 않는다.

### 연결 테스트

* 소스엔드포인트 콘솔로 이동한 후, '연결 테스트'를 누르면 복제 인스턴스와 DB가 잘 연동이 됐는지 확인할 수 있다.
* '상태'에 successful이라고 뜨면 정상적으로 잘 연동됐다고 볼 수 있다.

### Migration Task 생성

* 마이그레이션 테스크는 위에서 생성했던 엔드포인드와 복제 인스턴스를 선택해주면 되고, 마이그레이션 유형은 각자가 원하는 옵션을 선택하면 된다.

### 테이블 매핑

* 마이그레이션할 DB와 테이블을 입력하면 된다. 1개의 테이블만 선택할 수 있는 것은 아니고, 원하는 만큼 많이 입력해주면 된다.
* 변환규칙옵션을 활용하면 보다 유용한 전처리와 분석을 진행할 수 있다.

### 로그 활성화
* Task를 생성할 때, 모종의 이유로 태스크가 실패할 수 있다.
* CloudWatch를 통해서 로그를 기록하는 옵션을 활성화하는 것을 권장함 -> 에러가 발생했을 때, 원인을 정확히 파악할 수 있음

### Task 튜닝

* 데이터의 양이 매우 많은 테이블을 마이그레이션할 때는 Out of memory 에러가 발생할 수 있다. 
* 여기서, Replication instance의 메모리는 충분히 여유분이 있는데도 이러한 에러가 발생한 거라면, Full load tuning을 해줄 필요가 있다.

## 5. AWS Athena

* S3에 저장된 데이터를 기반으로 동작하는 서버리스 분산형 SQL엔진, 서버리스 인터랙티브 쿼리 서비스, Amazon S3에 저장된 데이터를 표준 SQL을 사용하여 쿼리할 수 있음
* S3에 저장된 데이터들을 ANSI SQL형식의 SQL로 조회
* 쉽게 말해, 사용자는 데이터를 불러오거나 인프라를 설정할 필요 없이 바로 쿼리할 수 있음

### 특징
* 서버리스: 인프라를 관리할 필요가 없으며, 사용한 쿼리 런타임에 대해서만 비용이 지불됨
* 표준 SQL 지원: SQL을 사용하여 데이터를 쿼리하고 분석
* 빠른 데이터 분석: S3에 저장된 데이터에 대해 빠르게 쿼리를 실행할 수 있어, 데이터 웨어하우스 없이도 가능
* 통합: 다양한 데이터 형식(JSON, CSV, Parquet 등)을 지원하며, Amazon QuickSight와 같은 시각화 도구와 용이

## 6. AWS QuickSight

* AWS에서 제공하는 serverless managed BI툴
* QuickSight를 사용하여 특정 데이터에 대한 시각화 대시보드를 생성하고 다른 사용자와 공유할 수 있다.

### 특징
1. **In-memory engine(SPICE) for fast performance**
* 다양한 데이터 소스로부터 데이터셋을 생성할 수 있다. 매번 데이터셋에 대한 데이터 조회 없이 일종의 중간 캐시에 해당하는 SPICE 저장소에 조회 결과를 저장하여 조회 속도를 향상시킬 수 있다.

2. **Built in data connectors**
* 구성 요소가 포함된 데이터 커넥터가 내장되어 있다.

3. **Integration with AWS services**
* AWS 서비스와 통합되기때문에 기존의 데이터 저장소에 대한 커넥터가 있어도 AWS 콘솔 내에서 접근하고 관리 할 수 있다.

4. **Fully cloud-based development**

### 데이터 쿼리 모드

#### 1. SPICE
QuickSight의 인메모리 데이터 엔진, 데이터를 QuickSight 내부의 고성능 메모리 엔진에 로드하여 쿼리를 매우 빠르게 실행할 수 있음
#### 특징
* 빠른 성능: 데이터가 SPICE 메모리에 저장되어 있어, 쿼리가 매우 빠르게 수행됩니다. 특히 복잡한 계산이나 대규모 데이터 세트에서 성능 이점이 크다.
* 비용 절감: SPICE 용량을 구입하여 데이터를 저장하고, 쿼리를 실행할 때 데이터 소스에 대한 비용이 발생하지 않으므로 비용 효율적
* 오프라인 액세스: 데이터가 SPICE에 로드되므로 데이터 소스가 오프라인 상태여도 대시보드를 계속 사용할 수 있음
* 최대 데이터 크기 제한: SPICE에 로드할 수 있는 데이터의 크기에 제한이 있으며, 더 많은 데이터가 필요하면 SPICE 용량을 추가 구매해야 합니다.
#### 예시사례
* 사용자가 자주 반복해서 조회하는 데이터에 대해 빠른 응답 시간이 필요한 경우.
* 데이터 소스에 자주 연결할 수 없거나 오프라인 상태에서도 대시보드가 필요할 때.
* 비용을 절감하면서도 대시보드 성능을 유지하고자 할 때.

#### 2. Direct Query
QuickSight에서 실시간으로 데이터 소스에 쿼리를 실행하는 방식
데이터를 미리 SPICE에 로드하지 않고, 사용자가 대시보드를 조회할 때마다 데이터 소스에 쿼리를 수행
#### 특징
* 실시간 데이터 접근: 항상 최신 데이터를 쿼리할 수 있습니다. 데이터가 자주 업데이트되거나 실시간 데이터가 필요한 경우 적합합니다.
* 대용량 데이터 지원: SPICE의 메모리 제한 없이 대규모 데이터를 쿼리할 수 있습니다.
* 데이터 소스 비용: 데이터 소스에 직접 쿼리를 수행하기 때문에, 데이터 소스에서 발생하는 비용(예: 데이터베이스 쿼리 비용)이 추가될 수 있습니다.
* 속도: 데이터 소스의 성능에 따라 쿼리 속도가 결정되므로, 데이터 소스의 처리 성능에 영향을 받을 수 있음.
#### 예시사례
* 데이터를 실시간으로 업데이트하고 최신 상태를 유지해야 하는 경우.
* SPICE 메모리 제한을 초과하는 대규모 데이터에 대한 분석이 필요한 경우
* 데이터 소스에 특화된 고급 기능(예: 데이터베이스 내 특정 함수나 프로시저)을 사용해야 할 때
