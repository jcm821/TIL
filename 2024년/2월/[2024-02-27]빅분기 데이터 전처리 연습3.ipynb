```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
# 56. 데이터를 로드하고 데이터 행과 열의 갯수 출력
csvFile = 'C:/Users/chunc/OneDrive/바탕 화면/Python연습/Credit Card/BankChurners.csv'
df = pd.read_csv(csvFile)
df.shape
```

```python
# 57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑
# Unknown : N
# Less than $40K : a
# $40K - $60K : b
# $60K - $80K : c
# $80K - $120K : d
# $120K + : e

mydict = {
    'Unknown': 'N',
    'Less than $40K': 'a',
    '$40K - $60K': 'b',
    '$60K - $80K': 'c',
    '$80K - $120K': 'd',
    '$120K +': 'e'
}

df['newIncome'] = df['Income_Category'].map(lambda x: mydict[x])
ans = df['newIncome']
ans.head()
```

```python
# 58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑
def myFunc(x):
    if x == 'Unknown':
        return 'N'
    elif x == 'Less than $40K':
        return 'a'
    elif x == '$40K - $60K':
        return 'b'
    elif x == '$60K - $80K':
        return 'c'
    elif x == '$80K - $120K':
        return 'd'
    else:
        return 'e'

df['newIncome'] = df['Income_Category'].apply(lambda x: myFunc(x))
# df['newIncome'] = df.Income_Category.apply(changeCategory)
ans = df['newIncome']
ans.head()
```

```python
# 59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의
# (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력
df['AgeState'] = df['Customer_Age'].map(lambda x: x // 10 * 10)
ans = df['AgeState'].value_counts().sort_values()
ans
```

```python
# 60. Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력
df['newEduLevel'] = df['Education_Level'].map(lambda x: 1 if 'Graduate' in x else 0)
ans = df['newEduLevel'].value_counts()
ans
```

```python
# 61. Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의
# newLimit 각 값들의 빈도수를 출력
df['newLimit'] = df['Credit_Limit'].map(lambda x: 1 if x >= 4500 else 0)
ans = df['newLimit'].value_counts()
ans
```

```python
# 62. Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의
# newState의 각 값들의 빈도수를 출력
def myFunc(x):
    if x['Marital_Status'] == 'Married' and x['Card_Category'] == 'Platinum':
        return 1
    else:
        return 0

df['newState'] = df.apply(lambda x: myFunc(x), axis = 1)
ans = df['newState'].value_counts()
ans
```

```python
# 63. Gender 컬럼값 M인 경우 male F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의
# 각 value의 빈도를 출력
def myGender(x):
    if x == 'M':
        return 'male'
    else:
        return 'female'

df['Gender'] = df['Gender'].apply(myGender)
ans = df['Gender'].value_counts()
ans
```

```python
# 64. 주가데이터를 로드하고 각 열의 데이터 타입 파악
csvFile = 'https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv'
df = pd.read_csv(csvFile)
df.info()
```

```python
# 65. Yr_Mo_Dy을 판다스에서 인식할 수 있는 datetime64타입으로 변경
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
ans = df['Yr_Mo_Dy']
ans.head()
```

```python
# 66. Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력
ans = df['Yr_Mo_Dy'].dt.year.unique()
ans
```

```python
# 67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다.
# 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의

def myYear(x):
    import datetime
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(myYear)
ans = df.head()
ans
```

```python
# 68. 년도별 각 컬럼의 평균값 구하기
ans = df.groupby(df['Yr_Mo_Dy'].dt.year).mean()
ans.head()
```

```python
# 69. weekday컬럼을 만들고 요일별로 매핑 ( 월요일: 0 ~ 일요일 :6)
df['weekday'] = df['Yr_Mo_Dy'].dt.weekday
ans = df['weekday'].head().to_frame()
ans
```

```python
# 70. weekday컬럼을 기준으로 주말이면 1 평일이면 0의 값을 가지는 WeekCheck 컬럼 생성
df['weekCheck'] = df['weekday'].map(lambda x: 1 if x in [5, 6] else 0)
ans = df['weekCheck'].head().to_frame()
ans
```

```python
# 71. 년도, 일자 상관없이 모든 컬럼의 각 달의 평균 산출
ans = df.groupby(df['Yr_Mo_Dy'].dt.month).mean()
ans
```

```python
# 72. 모든 결측치는 컬럼기준 직전의 값으로 대체하고 첫번째 행에 결측치가 있을 경우 뒤에있는 값으로 대체
df = df.fillna(method = 'ffill').fillna(method = 'bfill')
df.isnull().sum()
```

```python
# 73. 년도 - 월을 기준으로 모든 컬럼의 평균값
ans = df.groupby(df['Yr_Mo_Dy'].dt.to_period('M')).mean()
ans
```

```python
# 74. RPT 컬럼의 값을 일자별 기준으로 1차차분
ans = df['RPT'].diff(1)
ans.head()
```

```python
# 75. RPT와 VAL의 컬럼을 일주일 간격으로 각각 이동평균한값을 구하기
ans = df[['RPT', 'VAL']].rolling(window = 7).mean()
ans.head(10)
```

```python
# 76. 서울시 미세먼지 데이터를 이용
# 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경
# 서울시 제공데이터의 경우 0시가 24시로 표현
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
df.info()
```

```python
def myYear(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]

    if hour == '24':
        hour = '00:00:00'
        FinalDate = pd.to_datetime(date + ' ' + hour) + datetime.timedelta(days = 1)

    else:
        hour = hour + ':00:00'
        FinalDate = pd.to_datetime(date + ' ' + hour)

    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(myYear)
ans = df
ans.head()
```

```python
# 77. 일자별 영어요일 이름을 dayName 컬럼에 저장
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
ans = df['dayName']
ans.head()
```

```python
# 78. 일자별 각 PM10등급의 빈도수를 파악
ans1 = df.groupby(['dayName', 'PM10등급'], as_index = False).size()
ans2 = ans1.pivot(index = 'dayName', columns = 'PM10등급', values = 'size').fillna(0)
```
