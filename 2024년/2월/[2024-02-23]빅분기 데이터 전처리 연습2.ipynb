```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
# 44. 뉴옥 에어비엔비 데이터를 로드하고 상위 5개 컬럼을 출력
df = pd.read_csv('C:/Users/chunc/OneDrive/바탕 화면/Python연습/AirBnB/AB_NYC_2019.csv')
df.head()
```

```python
# 45. 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개 출력
# Ans = df.groupby('host_name').size()
ans = df['host_name'].value_counts().sort_index()
ans.head()
```

```python
# 46. 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임 생성. 빈도수 컬럼은 counts로 명명
ans = df.groupby('host_name').size().to_frame().rename(columns = {0:'counts'}).sort_values(by = 'counts', ascending = False)
ans.head()
```

```python
# 47. neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수 구하기
ans = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index = False).size()
ans.head()
```

```python
# 48. neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값 출력
ans = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index = False).size().groupby(['neighbourhood_group'], as_index = False).max()
ans
```

```python
# 49. neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값
ans = df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
ans
```

```python
# 50. neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값
ans = df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
ans
```

```python
# 51. neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균
ans = df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean()
ans
```

```python
# 52. neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하기
ans = df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean().unstack()
ans
```

```python
# 53. neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하고 nan 값은 -999값으로 할당
ans = df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean().unstack().fillna(-999)
ans
```

```python
# 54. 데이터중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값 출력
ans = df[df['neighbourhood_group'] == 'Queens'].groupby(['neighbourhood']).price.agg(['mean', 'var', 'max', 'min'])
ans.head()
```

```python
# 55. 데이터중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율 출력
ans = df[['neighbourhood_group', 'room_type']].groupby(['neighbourhood_group', 'room_type']).size().unstack()
ans.loc[:,:] = (ans.values / ans.sum(axis = 1).values.reshape(-1, 1))
ans
```
