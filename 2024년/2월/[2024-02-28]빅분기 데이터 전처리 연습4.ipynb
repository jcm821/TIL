```python
import pandas as pd
import numpy as np
```

```python
# 83. Indicator을 삭제하고 First Tooltip 컬럼에서 신뢰구간에 해당하는 표현 삭제
# 사망비율 통계 데이터 불러오기
csvFile = 'https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv'
df = pd.read_csv(csvFile)
df.head()
```

```python
# 데이터 정보 확인
df.info()
```

```python
# Indicator 삭제
df.drop('Indicator', axis = 1, inplace = True)

# First Tooltip 컬럼에서 신뢰구간에 해당하는 표현 삭제
# df['First Tooltip'] = df['First Tooltip'].str.split('[').str[0].astype('float')
df['First Tooltip'] = df['First Tooltip'].map(lambda x: float(x.split('[')[0]))
df.head()
```

```python
# 84. 년도가 2015년 이상, Dim1이 Both sexes인 케이스만 추출
target = df[(df['Period'] >= 2015) & (df['Dim1'] == 'Both sexes')]
ans = target
ans.head()
```

```python
# 85. 84번 문제에서 추출한 데이터로 아래와 같이 나라에 따른 년도별 사망률을 데이터 프레임화
ans = target.pivot_table(index = 'Location', columns = 'Period', values = 'First Tooltip')
ans.head()
```

```python
# 86. Dim1에 따른 년도별 사망비율의 평균
ans = df.pivot_table(index = 'Dim1', columns = 'Period', values = 'First Tooltip', aggfunc = 'mean')
ans.iloc[:, :4]
```

```python
# 87. 올림픽 메달리스트 데이터를 이용하여 데이터에서 한국 KOR 데이터만 추출
# 데이터 불러오기
csvFile = 'https://raw.githubusercontent.com/Datamanim/pandas/main/winter.csv'
df = pd.read_csv(csvFile)
df.head()
```

```python
# 데이터 정보 확인
df.info()
```

```python
# 한국 KOR 데이터만 추출
new_df = df[df['Country'] == 'KOR']
new_df.head()
```

```python
# 88. 한국 올림픽 메달리스트 데이터에서 년도에 따른 medal 갯수를 데이터프레임화
ans = new_df.pivot_table(index = 'Year', columns = 'Medal', aggfunc = 'size').fillna(0)
ans.head()
```

```python
# 89. 전체 데이터에서 sport종류에 따른 성별수 구하기
ans = df.pivot_table(index = 'Sport', columns = 'Gender', aggfunc = 'size')
ans
```

```python
# 90. 전체 데이터에서 Discipline종류에 따른 Medal수를 구하기
ans = df.pivot_table(index = 'Discipline', columns = 'Medal', aggfunc = 'size')
ans
```

```python
# 91. 사망비율 데이터를 이용하여 df1과 df2 데이터를 하나의 데이터 프레임으로 합치기
# 데이터 불러오기
csvFile = 'https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv'
df = pd.read_csv(csvFile, index_col = 0)
df.head()
```

```python
# 데이터 분할
df1 = df.iloc[:4, :]
df2 = df.iloc[4:, :]

display(df1)
display(df2)
```

```python
# 두 데이터를 하나의 데이터프레임으로 병함
total = pd.concat([df1, df2], axis = 0)
ans = total
ans
```

```python
# 다른 조건으로 데이터를 분할
df3 = df.iloc[:2, :4]
df4 = df.iloc[5:, 3:]

display(df3)
display(df4)
```

```python
# 92. df3과 df4 데이터를 하나의 데이터 프레임으로 합치기 (둘다 포함하고 있는 년도에 대해서만 고려)
ans = pd.concat([df3, df4], axis = 0, join = 'inner')
ans
```

```python
# 93. df3과 df4 데이터를 하나의 데이터 프레임으로 합치기 (모든 컬럼을 포함하고, 결측치는 0으로 대체)
ans = pd.concat([df3, df4], axis = 0, join = 'outer').fillna(0)
ans
```

```python
# 다른 조건으로 데이터를 분할
df5 = df.T.iloc[:7, :3]
df6 = df.T.iloc[6:, 2:5]

display(df5)
display(df6)
```

```python
# 94. df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합치기 (Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력)
ans = pd.merge(df5, df6, on = 'Algeria', how = 'inner')
# ans = df5.merge(df6, on = 'Algeria', how = 'inner')
ans
```

```python
# 95. df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합치기 (Algeria컬럼을 key로 하고 합집합으로 병합)
ans = pd.merge(df5, df6, on = 'Algeria', how = 'outer')
# ans = df5.merge(df6, on = 'Algeria', how = 'outer')
ans
```
