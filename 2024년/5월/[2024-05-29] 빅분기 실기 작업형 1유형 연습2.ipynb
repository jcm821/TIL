## 작업형 1유형

```python
# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv')
df.head()
```

```python
# 72. 모든 결측치는 컬럼기준 직전의 값으로 대체하고 첫번째 행에 결측치가 있을경우 뒤에있는 값으로 대체하라
df = df.fillna(method = 'ffill').fillna(method = 'bfill')
```

```python
# 결측값 확인
df.isnull().sum()
```

```python
# 데이터 타입 변환
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
```

```python
def fix_century(x):
    import datetime
    
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))


df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)

result = df.head(4)
result
```

```python
# 73. 월을 기준으로 모든 컬럼의 평균값을 구하여라
result = df.groupby(df['Yr_Mo_Dy'].dt.to_period('M')).mean()
result.head()
```

```python
# 74. RPT 컬럼의 값을 일자별 기준으로 1차차분하라
result = df['RPT'].diff()
result.head()
```

```python
# 75. RPT와 VAL의 컬럼을 일주일 간격으로 각각 이동평균한값을 구하여라
result = df[['RPT', 'VAL']].rolling(7).mean()
result.head(9)
```

```python
# 76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 
# 서울시의 제공데이터의 경우 0시가 24시로 표현된다
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')

def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]

    if hour == '24':
        hour = '00:00:00'

        FinalDate = pd.to_datetime(date + ' ' + hour) + datetime.timedelta(days = 1)

    else:
        hour = hour + ':00:00'
        FinalDate = pd.to_datetime(date + ' ' + hour)

    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
result = df
result.head()
```

```python
# 77. 일자별 영어요일 이름을 dayName 컬럼에 저장하라
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
result = df['dayName']
result.head()
```

```python
# 78. 일자별 각 PM10등급의 빈도수를 파악하라
result1 = df.groupby(['dayName', 'PM10등급'], as_index = False).size()
result2 = result1.pivot(index = 'dayName', columns = 'PM10등급', values = 'size').fillna(0)
```
