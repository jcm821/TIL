## 파이썬 메소드를 활용예제(os, shutil)


```python
import os, shutil
```

```python
# 1번예제: 파일생성(touch로/ content디렉토리 내에 aa/txt)
# !touch /content/aa.txt도 가능
!pwd
!touch aa.txt
```

```python
# path메소드 이용하여 빈 파일 생성
from pathlib import Path
Path('aa.txt').touch()
os.listdir('.')
```

```python
# 2번예제:디렉토리 생성하기 (test3)
import os
os.makedirs('./test3', exist_ok = True)
```

```python
# test3디렉토리 생성여부 확인
os.listdir('.')
```

```python
# 3번예제: test3디렉토리로 복사하기
import shutil
srcFile = './aa.txt'
dstFile = './test3/aa.txt'
shutil.copy(srcFile, dstFile)
```

```python
# 복사가 잘 되었는지 여부 확인
os.listdir('/content/test3')
```

```python
# 4번예제: test3디렉토리안에서 aa.txt파일을 bb.txt파일로 복사하기
import shutil
srcFile = './test3/aa.txt'
dstFile = './test3/bb.txt'
shutil.copy(srcFile, dstFile)
```

```python
# 복사여부 확인
os.listdir('/content/test3')
```

```python
# 5번예제: bb.txt파일을 /content/test4/디렉토리로 복사하기
# copy를 사용한 방법
# '.'대신 content를 그대로 적용해도 상관없다
import os
import shutil
os.makedirs('/content/test4', exist_ok = True)
srcFile = '/content/test3/bb.txt'
dstFile = '/content/test4/bb.txt'
shutil.copy(srcFile, dstFile)
```

```python
# 복사여부 확인
os.listdir('/content/test4')
```

```python
# 5번예제: bb.txt파일을 /content/test4/디렉토리로 복사하기
# move를 사용한 방법
import os
import shutil
os.makedirs('./content/test4', exist_ok = True)
srcFile = './test3/bb.txt'
dstFile = './content/test4/bb.txt'
shutil.move(srcFile, dstFile)
```
