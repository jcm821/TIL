따라서 원본 데이터를 유지시키고, 새로운 변수에 복사할 때는 copy()를 사용한다.

```python
df = pd.read_csv('C:/Users/chunc/Desktop/취업캠프/python-20230220T004050Z-001/python/korean-idol.csv')
df.head()
```

```python
# python에서 얕은 복사(주소값만 복사), 깊은 복사(deep copy)가 있는데
# copy()함수를 사용하면 깊은 복사가 실행된다
copy_df = df.copy()
```

```python
id(df)
```

```python
# 일반변수는 얕은 복사를 신경쓰지 않아도 된다
# a변수를 복사해서 b에 가져다 놓으면 일단 id는 동일
# 만약 a나 b를 변경하는 순간 id값이 달라진다(그 순간 deep copy가 이루어진다)
a = 15
b = a
a += 1
print(id(a))
print(id(b))
```

### 3-3. var, variance(분산) / std, standard deviation(표준편차)

* 분산과 표준편차는 데이터가 평균으로부터 얼마나 떨어져 있는지 정도를 나타낸다.
* 분산은(데이터 - 평균) ** 2 을 모두 합한 값 / n
* 표준편차는 분산의 루트

```python
import numpy as np
```

```python
data_01 = np.array([1, 3, 5, 7, 9])
data_02 = np.array([3, 4, 5, 6, 7])
```

```python
# data_01 표준편차 구한식
np.sqrt(((1 - 5)**2 + (3 - 5)**2 + (5 - 5)**2 + (5 - 5)**2 + (9 - 5)**2)/ 5)
```

```python
data_01.mean()
```

```python
data_02.mean()
```

```python
# data_01의 분산, data_02의 분산
data_01.var(), data_02.var()
```

```python
np.sqrt(data_01.var()), np.sqrt(data_02.var())
```

```python
# data_01의 표준편차, data_02의 표준편차
data_01.std(), data_02.std()
```

주로 표준편차(std)를 많이 사용하며, 데이터가 평균으로부터 얼마나 퍼져있는지 정도를 나타내는 지표이다

### 8-2. DataFrame 병합하기(merge)

이전에 봤던 concat과 merge는 단순 합치는 목적과 특정 기준(index)으로 합치느냐에 따라 용도가 다르다.
* concat: row나 column 기준으로 단순하게 이어 붙이기
* merge: 특정 고유한 키(unique id)값을 기준으로 병합하기

```python
df = pd.read_csv('C:/Users/chunc/Desktop/취업캠프/python-20230220T004050Z-001/python/korean-idol.csv')
df2 = pd.read_csv('C:/Users/chunc/Desktop/취업캠프/python-20230220T004050Z-001/python/korean-idol-2.csv')
```

```python
df.head()
```

```python
df2.head()
```

df와 df2는 '이름'이라는 column이 겹친다.<br>
따라서, 우리는 '이름'을 기준으로 두 데이터프레임을 병합할 수 있다.

**pd.merge(left, right, on='기준column', how='left')**

* left와 right는 병합할 두 DataFrame을 대입합니다.
* on 에는 병합의 기준이 되는 column을 넣어 줍니다.
* how 에는 'left', 'right', 'inner', 'outer' 라는 4가지의 병합 방식중 한가지를 택합니다.

```python
# df2에서 5개의 데이터를 고의적으로 drop을 실시해준다
df_right = df2.drop([1, 3, 5, 7, 9], axis = 0)
df_right
```

```python
df_right = df_right.reset_index(drop = True)
df_right
```

```python
# 이름이나 데이터를 보고 데이터프레임을 합쳐주는것이 아니라
# 기계적으로 같은 행이나 열을 맞춰서 merge해주는 함수
pd.concat([df, df_right], axis=1)
```

```python
df
```

병합할 2개의 DataFrame의 행의 갯수가 맞지 않는다. <br>
'left'옵션을 부여하면, left DataFrame에 키 값이 존재하면 해당 데이터를 유지하고, 병합한 right DataFrame의 값은 NaN이 대입된다.

```python
pd.merge(df, df_right, on = '이름', how = 'right')
```

위의 예제에서, right를 기준으로 병합했기 때문에 DataFrame사이즈가 줄어든다.

#### inner, outer방식

* inner 방식은 두 DataFrame에 모두 키 값이 존재하는 경우만 병합합니다.
* outer 방식은 하나의 DataFrame에 키 값이 존재하는 경우 모두 병합합니다.
* outer 방식에서는 없는 값은 NaN으로 대입됩니다.

```python
df
```

```python
df_right
```

```python
# 2개의 데이터프레임을 합쳤을때 두 데이터 프레임중에 결측치가 있는 행은 삭제
# 오른쪽 데이터프레임의 '이름' column의 결측치가 있는 경우
pd.merge(df, df_right, on = '이름')
# how = 'inner'
```

```python
# 2개의 데이터프레임중에 하나라도 데이터가 존재하면 그 행은 합쳤을때 존재한다.
pd.merge(df, df_right, on='이름', how='outer')
```
