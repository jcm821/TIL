## 데이터프레임 합치기 연습

```python
# 합칠 데이터 불러오기
import pandas as pd
csvPath = 'C:/Users/chunc/Desktop/취업캠프/python-20230220T004050Z-001/python/01. CCTV_in_Seoul.csv'
xlsPath = 'C:/Users/chunc/Desktop/취업캠프/python-20230220T004050Z-001/python/01. population_in_Seoul.xls'
```

```python
df = pd.read_csv(csvPath)
df2 = pd.read_excel(xlsPath, header = 2, usecols = 'B, D, I, J, N')
```

```python
# 데이터 확인
df
```

```python
# df2의 이름변경 해주기
df2.columns = ['구별', '서울전체', '한국인여자', '등록외국인', '고령자']
df2
```

## 방법1: concat함수 사용

```python
# concat함수 이용하여 데이터프레임 합치기
# axis = 0
concat_df1 = pd.concat([df, df2], sort = False, axis = 0)
concat_df1
```

```python
# axis = 1
concat_df2 = pd.concat([df, df2], sort = False, axis = 1)
concat_df2
```

```python
# 인덱스 번호 초기화
concat_df1.reset_index(drop = True)
```

```python
concat_df2.reset_index(drop = True)
```

```python
# 컬럼명이 중복일때 바꿔주는 방법
# 방법1: 합친 후 컬럼명을 바꿔주어 컬럼을 삭제해준다
# 방법2: 합치기 전 이름을 달리하여 저장한 후 컬럼을 삭제해준다
concat_df1.drop('구별', axis = 1)
```

```python
concat_df2.drop('구별', axis = 1)
```

## 방법2: merge함수 이용하기

```python
df = pd.read_csv(csvPath)
df2 = pd.read_excel(xlsPath, header = 2, usecols = 'B, D, I, J, N')
```

```python
df2.columns = ['구별', '서울전체', '한국인여자', '등록외국인', '고령자']
df2
```

```python
# merge함수: inner조인
merge_df = pd.merge(df, df2, left_on = '기관명', right_on = '구별', how = 'inner')
merge_df
```

```python
# 중복컬럼 삭제
merge_df.drop('구별', axis = 1, inplace = True)
```

```python
merge_df
```

```python
# merge함수: outer조인
merge_df2 = pd.merge(df, df2, left_on = '기관명', right_on = '구별', how = 'outer')
merge_df2
```

```python
# 중복칼럼 제거
merge_df2.drop('구별', axis = 1, inplace =True)
```

```python
merge_df2
```
