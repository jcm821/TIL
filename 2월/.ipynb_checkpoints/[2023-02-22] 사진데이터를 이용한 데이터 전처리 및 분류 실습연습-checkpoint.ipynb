## train 데이터로 사진을 각 폴더에 맞게 분류하기

### 1. 이미지 저장 폴더 데이터를 문자열 변수에 생성

```python
# 이미지가 저장된 폴더 데이터를 문자열 변수에 생성
# src_path = 'C:\\Users\\chunc\\PycharmProjects\\pythonEx1\\train\\train'와 같이 작성해도 무방하다
srcPath = 'C:/Users/chunc/PycharmProjects/pythonEx1/train/train/'
```

### 2. 파일목록 불러오기 

```python
import os, shutil
# srcPath의 파일목록을 불러온다
filename = os.listdir(srcPath)
print(filename)
```

```python
print(srcPath)
```

### 3. 파일 개수 확인

```python
# 폴더의 파일 개수 확인
print(len(filename))
```

### 4. 나누고자 하는 폴더 생성

```python
# dog, cat폴더 각각 생성
dirname1 = srcPath + 'dog'
dirname2 = srcPath + 'cat'
os.makedirs(dirname1, exist_ok = True)
os.makedirs(dirname2, exist_ok = True)

```

### 5. JPG파일만 추출하기

glob: 특정 파일만 출력할 수 있다

```python
# glob.glob('*.jpg') 
# jpg파일만 추출한다
```

```python
# glob모듈을 사용해서 jpg파일만 추출(찾고자하는 확장자의 파일만 필터링 할 수 있음)
import glob
jpgFile = glob.glob(srcPath+'*.jpg')
print(jpgFile)
```

```python
# jpg파읾의 전체길이 확인
print(len(jpgFile))
```

### 6. 추출된 파일을 확인 후, 각 폴더로 데이터 이동시키기

```python
# base모듈을 사용하여 전체 경로를 삭제하고 파일명만 보고자 한다
# 'cat.0.jpg'
for src in jpgFile:
    # basename이라는 메서드로  파일전체 경로를 삭제하고
    # 파일명(cat.0.jpg)만 남겨주는 함수이다
    
    # cat.0.jpg의 파일명을 baseFileName이라는 변수에 저장
    baseFileName = os.path.basename(src)
    
    
    # 파일명에서 'cat', 'dog'만 추출한다
    # cat, dog은 category로, '0'은 첫번째 _에, jpg는 두번째 _에 할당된다
    # 언더바를 통해 분류에 필요없는 부분은 제외하고 추출해낸다
    category, _, _ = baseFileName.split('.')
    
    # srcPath - 'C:/Users/chunc/PycharmProjects/pythonEx1/train/train/'
    # category - cat
    # baseFileName - cat.0.jpg
    # 다 합쳐서 'C:/Users/chunc/PycharmProjects/pythonEx1/train/train/cat/cat.0.jpg'를 dst변수에 저장
    dst = srcPath + category + '/' + baseFileName
    
    # 파일을 이동한다.
    # src = 'C:/Users/chunc/PycharmProjects/pythonEx1/train/train/cat.0.jpg'에서
    # dst = 'C:/Users/chunc/PycharmProjects/pythonEx1/train/train/cat/cat.0.jpg'으로 이동한다
    shutil.move(src, dst)
    
    
    # 숙달을 위한 코드실햏 시에는 shutil.copy함수를 사용하는 것이 용이하다
```

이후 디렉토리를 확인해보면, dog, cat폴더에 각각 사진이 이동된 것을 알 수 있다.

#### 경로명 분류하기

os.path.basename(filename) - 파일명만 추출한다<br>
os.path.dirname(filename) - 디렉토리 경로를 추출한다<br>
os.path.split(filename) - 경로와 파일명을 분리한다<br>
os.path.splitdrive(filename) - 드라이브명과 나머지 분리 (MS Windows의 경우)<br>
os.path.splitext(filename) - 확장자와 나머지 분리한다<br>
